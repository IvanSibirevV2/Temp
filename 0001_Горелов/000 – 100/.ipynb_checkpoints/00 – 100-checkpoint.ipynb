# ПРЕДИСЛОВИЕ
Разработка приложений с графическим интерфейсом, кроме написания кода, включает в себя также проектирование и разработку интерфейса пользователя. Под интерфейсом пользователя понимаются окна приложения, их элементы управления и организация взаимодействия с пользователем. Качество интерфейса, его эргономика, художественное оформление и т.д. определяют качество и привлекательность приложения. Вспомним, сколько раз менялся интерфейс таких Интернет-ресурсов, как портал государственных услуг (https://www.mos.ru/) или портал Финансового университета (www.fa.ru). Разработчики этих порталов предпринимают много усилий для того, чтобы сделать свой продукт удобным. __Наиболее эффективные__ и развитые подходы к разработке Windows-приложений предлагает платформа Framework .NET и ее среда разработки Visual Studio. В этой платформе реализованы две технологии разработки интерфейса пользователя: Windows Forms и Windows Presentation Foundation (WPF). WPF является новой технологией, основанной на XML. Старая традиционная технология практически не требует знания языка XML, поэтому она проще новой. __В данном учебнике__ излагаются основы разработки Windows приложений по технологии Windows Forms. Эта технология представлена типами пространства имен System. Windows.Forms, к которым относятся элементы управления, компоненты, диалоговые окна и другие типы. __Среда Visual Studio__ позволяет проектировать окна приложения в режиме визуальной разработки, при этом во многих случаях программисту не требуется разрабатывать код для корректировки интерфейса. Программист направляет свои усилия лишь на разработку обработчиков событий. При этом разработчик все равно должен знать назначение, свойства и поведение элементов интерфейса и использовать наиболее подходящие элементы для решения задачи. Количество элементов управления и компонентов превышает 60. Подробно изучить их в одном учебнике не представляется возможным. В учебнике излагаются лишь наиболее часто используемые свойства и методы этих элементов, что в большинстве случаев не препятствует созданию приложения с развитым эргономичным интерфейсом. Все типовые применения элементов управления и компонентов проиллюстрированы примерами на языке С#.

# 1. ВЗАИМОДЕЙСТВИЕ ПОЛЬЗОВАТЕЛЯ С ПРОГРАММОЙ
## 1.1. Типы программ
Существуют два типа прикладных программ (ПП), различающихся способом взаимодействия с Windows и свойствами создаваемого окна. Это консольные приложения и приложения с графическим интерфейсом. Для консольных приложений операционная система Windows создает окно, являющееся моделью окна MS DOS с текстовым режимом. Консоль состоит из окна и клавиатуры. Манипулятор мышь не поддерживается. __С появлением Windows__ основными приложениями стали приложения с графическим интерфейсом. Такие приложения будем называть Windows-приложениями. __Консольная программа инициирует__ взаимодействие с пользователем с помощью запросов на выполнение того или иного метода. На рис. 1.1 показана схема взаимодействия на примере ввода с клавиатуры с помощью метода System. Console. ReadLine. Консольная программа путем выполнения методов консольного ввода (запросов) определяет моменты времени и последовательность взаимодействия программы с пользователем. __Однако Windows-приложения__ работают не так. Пользователь сам выбирает, какой элемент управления, размещенный на окне, он хочет использовать. Взаимодействие программы с пользователем берет на себя Windows. Прикладная программа теперь не запраши- васт с помощью функций ввод данных, вместо этого она ждет, ког- да Windows известит программу о том, что пользователь выполнил какое-то действие. И к этим действиям относится не только ввод данных, но и огромное количество других действий, например выборка пункта меню. Этот процесс проиллюстрирован на рис. 1.2.

Рис. 1.1. Взаимодействие консольной программы с пользователем, управляемое запросами (функциями)

Рис. 1.2. Взаимодействие Windows-программы с пользователем, управляемое событиями

Программа ожидает сообщения, посылаемого ей операционной системой Windows. Сообщение передается в программу посредством вызова специальной функции, которая также выполняется системой Windows. После получения сообщения программа, как правило, выполняет запрограммированные действия. Взаимодействие программы и операционной системы, базирующееся на сообщениях, определяет схему построения всех Windows-программ. __B Windows существует__ множество различных типов сообщений. Например, каждый раз при щелчке мыши в пределах окна программы ей будет направлено сообщение о нажатии кнопки мыши. Клавиатурное сообщение приходит, если нажата клавиша на клавиатуре в то время, когда окно программы имеет фокус ввода (т.е. ввод с клавиатуры направляется окну). Когда программа начинает работать, сообщения к ней приходят неупорядоченным образом, т.е. программа никогда «не знает», какое сообщение будет следующим. Таким образом, Windows-программы напоминают программы, работающие по прерываниям.

# 1.2. Цикл обработки сообщений
Как отмечалось ранее, Windows взаимодействует с программой, посылая ей сообщения. При этом существует системная очередь сообщений — одна на всю систему и по одной очереди на каждый поток. На рис. 1.3 показана схема обработки событий, реализованная в Windows. __Системная очередь__, накапливая и перераспределяя сообщения по очередям потоков, решает задачи синхронизации. Все приложения Windows должны запускать цикл обработки сообщений. __В программах__ на языке С++ цикл реализует разработчик в функции WinMain. В С#-программах такой цикл создается автоматически с помощью метода Application.Run. Этот метод в качестве араметра получает ссылку на главное окно программы. Цикл обаботки сообщений выбирает сообщения из очереди сообщений приложения и направляет их обратно в Windows, которая затем вызывает оконную функцию программы, передавая ей указанные сообщения в качестве параметров. В программах на языке С+ оконная функция чаще всего имеет имя WndProc. Описанный способ взаимодействия может показаться излишне сложным, однако одной из причин реализации такой схемы является то, что программа должна возвращать управление планировщику Windows для распределения времени центрального процессора и его оптимальной загрузки.

    float curs = 50.0f;

    private void button1_Click(object sender, EventArgs e)

        {

            textBox1.Focus();

    dollars = rubli / curs;
    textBox2.Text = dollars.ToString();
}

Подробнее о добавлении обработчиков событий. Не вдаваясь в подробности, отметим, что созданию обработчика события предшествует этап подписки на это событие. Для того чтобы подписаться на требуемое событие какого-либо элемента управления и создать обработчик события, в общем случае необходимо: __Во-первых__, узнать, какое имя имеет событие; __Во-вторых__, выяснить, какой делегат представляет событие, и по делегату определить формат заголовка обработчика события; __В-третьих__, подписаться на событие, то есть добавить обработчик к событию; __В четвёртых__, создать пустой обработчик события с требуемыми параметрами; __В пятых__, разработать тело обработчика события. __Перечисленные выше шаги__ соответствуют модели событийного управления, реализованного в Windows. Выполнение их вручную - достаточно сложный для начинающего разработчика процесс, однако среда Visual Studio первые четыре шага выполняет автоматически в ответ на Ваши действия мышкой.     textBox1.TextChanged += button1_Click;

# 2.2. Подробнее о делегатах и событиях
Если для управления событиями в приложениях с графическим интерфейсом Вам достаточно приведенной выше информации, то данный раздел можно пропустить. Ниже речь пойдет об углубленном описании понятий делегат и событие применительно к Windows-приложениям.
# ДЕЛЕГАТЫ
В событийной модели используются групповые (многоадресатные) делегаты. Делегат — это класс, содержащий данные о том, какую сигнатуру должен иметь метод, указываемый в экземпляре этого класса. В экземпляре можно определить только тот метод, у которого тип возвращаемого значения (в случае событийной модели — void) и список параметров совпадают с соответствующими элементами объявления делегата. Информация, описывающая заголовок метода, называется сигнатурой. В экземпляре делегата хранится ссылка на конкретный метод, поэтому вызов на выполнение экземпляра равносилен вызову метода. Общая форма объявления группового делегата:
    
    delegate void имя_делегата (список_параметров);

Пусть групповой делегат MyDlg объявлен следующим образом: delegate void MyDig(int a, bool b); тогда в экземпляре этого делегата в качестве параметра можно указать имя следующего метода:

    void Method (int sum, bool fl){/*...*/}

Страница 32

---------

Это допустимо, так как сигнатура метода Method соответствует делегату MyDIg. Экземпляр делегата для метода Method создается так же, как и любой объект, например

    MyDig Handler = new MyDIg (Method);

В качестве параметра конструктора указывается вызываемый метод. Поскольку этот метод будет вызывать система, то такой метод называется методом обратного вызова. Как было указано ранее, существует более простой способ создания экземпляра делегата, когда указывается только имя метода, например:

    MyDIg Handler = Method;

# СОБЫТИЯ
Событийная модель в CLR строится на основе групповых делегатов. В событийной модели применяется схема издатель-подписчик. В этой модели в различных источниках используются термины: издатель, подписчик, уведомление, подписка, регистрация, открепление и др. В данном учебнике для упрощения материала, по возможности, будут использоваться понятия из языка С#. __На уровне Windows__ аппаратные события возникают от клавиатуры, мышки и таймера. В классах различных элементов управления присутствуют методы, создающие программные события, например «потеря элементом фокуса ввода». В языке С# событие представлено членом класса, который объявляется с помощью ключевого слова event. Синтаксис наиболее распространенной формы объявления события имеет следующий вид:

    event имя делегата имя_события;

Из этого формата следует, что событие объявляется на основе созданного ранее группового делегата. Приведем пример объявления события Click класса Button:

    public event EventHandler Click;
    
Страница 33

------

Из примера следует, что в элементе управления типа Button определено событие Click, которому может быть присвоен экземпляр делегата System.EventHandler с соответствующим по сигнатуре обработчиком события. Эта операция называется подпиской на событие. Например, по двойному щелчку элемента button1 среда Visual Studio сгенерирует (в Form1.Designer.cs) следующую подписку:
    
    button1.Click += new System.EventHandler(button1_Click);

Также в Form1.cs будет сгенерирован пустой обработчик button1_Click. Сигнатура этого обработчика будет соответствовать делегату System.EventHandler. В каждом элементе управления определено множество событий. В визуальном режиме (в окне «Свойства») разработчик может выбрать событие, возникновение которого необходимо обрабатывать, и двойным щелчком или путем выборки старого обработчика выполнить подписку. Иногда разработчику требуется выполнить подписку на событие в том месте кода, где ему это необходимо. В этом случае он может использовать упрощенный синтаксис, например, для вышеприведенного примера такая подписка будет выглядеть следующим образом:

    button1.Click += button1_Click;

Страница 34

------------

Если обработчик события имеет маленький размер, то его тело можно перенести в лямбда-выражение, например:

    button1.Click += (sender, args) => Close();

В этом случае анонимный обработчик события будет создан автоматически. Составные операции «+=» и «-=» позволяют добавлять ссылки на методы в экземпляр группового делегата и соответственно удалять их оттуда. Вызов события на выполнение приводит к последовательному вызову всех методов, ссылки на которые хранит экземпляр группового делегата. Вызов экземпляра также приводит к выполнению методов списка. Таким образом, член event — это дополнительный слой программного обеспечения. Такие методы называются обработчиками событий. __В классах__, представляющих элементы управления, определение событий и их вызов на выполнение осуществляется в ме тодах, имя которых начитается с On. Так, определение события Click и вызов его обработчиков выполняется методом OnClick, а события Enter — методом OnEnter. Методы On... вызываются либо программно, либо при возникновении вложенного события более низкого уровня. Если переопределить или заместить метод On..., добавив свою обработку события, которое он вызывает, то необходимость в подписке на это событие отпадает. Приведем пример. Пусть необходимо выдать сообщение в случае щелчка формы Form1, не подписываясь на событие Click:

    protected override void OnClick(EventArgs e)
    {
        MessageBox.Show("Для формы не определена обработка события Click!*);
        base.OnClick(e);
    }

Есть одно ограничение: все методы On... защищены (Protected) и вызвать их можно только из производного класса, которым, например, является класс Form1. Однако необходимость использования методов ОП.. возникает редко, __И последнее__. Если Вы будете реализовывать свою событийную модель в приложении, то можно отказаться от членов event и использовать в классе только экземпляры групповых делегатов. Причина, по которой для классов был введен член event, для читателя неочевидна. Разные авторы часто указывают на повышение безопасности. В частности, есть мнение, что дело заключается в том, что экземпляр группового делегата (как и многие объекты) имеет методы, с помощью которых пользователь может повлиять на чужие ссылки на методы. Событие же не является объектом и поэтому не имеет каких-либо методов. Удалим из разрабатываемого нами приложения лишние подписки и методы, созданные в качестве примеров. Рассмотрим файловый состав и код приложения.

# 3. ОБЩИЕ СВЕДЕНИЯ ОБ ЭЛЕМЕНТАХ УПРАВЛЕНИЯ
Элемент управления - это отдельный элемент пользовательского интерфейса, отображающий данные и/или принимающий ввод данных. В программе элементы управления являются объектами, обеспечивающими взаимодействие пользователя с программой. Каждый элемент управления представлен классом. __В среде__ разработки Visual Studio существует несколько шаблонов, по которым создаются проекты приложений с графическим интерфейсом. Для каждого шаблона используются свои элементы управления. Наиболее распространенным шаблоном для разработки Windows-приложений является Windows Forms. Альтернативным и более новым вариантом является шаблон WPF со своими элементами управления. Также свои элементы управления имеют шаблоны для ASP.NET и ADO. NET. Кроме того, система программирования С# предоставляет разработчикам средства для создания пользовательских элементов управления. __Все элементы__ проекта Windows Forms представлены в Visual Studio в окне «Панель элементов», которое можно отобразить с помощью пункта меню «Вид». Элементы представлены классами пространства имен System.Windows.Forms. Насчитывается более шестидесяти элементов, которые делятся на пять групп (Visual Studio 2015): __Во-первых__, Стандартные элементы управления (21 элемент); __Во-вторых__, Контейнеры (6); __В-третьих__, Меню (5); __В четвёртых__, Данные (5); __В пятых__, Компоненты (14); __В Шестых__, Печать (5); __В Седьмых__, Диалоговые окна (5); __В Восьмых__, Взаимодействие WPF (1).

Почти все элементы управления наследуют класс Component, поэтому они являются компонентами. Класс Component предо ставляет базовую реализацию интерфейса IComponent и делает возможным совместное использование объектов разными прило жениями. __При этом__ многие элементы прямо или косвенно наследуют класс Control, который создает соответствующее изображение на форме. Элементы используются для создания пользовательского интерфейса, поэтому эти элементы называют элементами управления. Класс Control реализует базовые функциональные возможности для всех элементов управления. Рис. 3.1. Форма и панель компонентов

Некоторые элементы управления, которые наследуют класс Control, отображаются на панели компонентов, но не отображаются на форме. В форме отображаются только коллекции этих контейнеров. __Кроме элементов__ управления, пространство имен System.Windows.Forms предоставляет классы, которые не являются производными от класса Control, но также обеспечивают визуальные функции для приложений Windows. __Ряд элементов__ управления являются контейнерами, предназначенными для хранения других элементов управления. К таким контейнерам относятся Panel, GroupBox и др. Главным контейнером элементов управления является форма, представленная классом Form. Все элементы управления, размещенные на форме, сохраняются в её коллекции. Не отображаемые элементы, то есть компоненты, в коллекцию формы не добавляются. __B Visual Studio__ все элементы пользовательского интерфейса перечислены в окне «Панель элементов». ### 4.1. Свойства класса Control
Как уже не раз подчеркивалось, класс Control является базовым для всех элементов управления и форм. Ниже рассмотрим только наиболее часто используемые члены класса Control. Внешнии вио элемента управления определяют следующие свойства:
- Text — текст, отображаемый в элементе управления.
- Font — шрифт текста.
- ForeColor — цвет переднего плана элемента управления, то есть текста.

- BackGroundImageLayout — способ отображения графического образа в элементе управления. Возможные значения: Center, None, Stretch, Tile или Zoom. Tile является значением по умолчанию.

# 5. ВНЕДРЕНИЕ ИЗОБРАЖЕНИЯ В СБОРКУ

Многие элементы управления, имеющие поверхность, например Button, имеют свойства, позволяющие либо использовать изображение в качестве фона, либо нарисовать его поверх фона. Среда Visual Studio оформляет изображение в виде так называемого ресурса. Не вдаваясь в подробности, можно сказать, что ресурс — это данные, внедряемые в сборку. После внедрения изображения в сборку, исходный графический файл станет ненужным. __Когда разработчик__ определяет файл с изображением для элемента управления, среда Visual Studio предлагает выбрать, в какое место поместить ресурс-изображение (точнее, с каким resx-файлом ассоциировать файл ресурса) (см. рис. 5.1). Вы можете выбрать либо «Локальный ресурс», либо «Файл ресурсов проекта». Изображение в обоих случаях в виде ресурса будет внедрено в сборку, однако в последнем случае в проекте будет создана папка Resources и в неё будет скопирован файл с изображением. Этот файл всегда будет доступен разработчику для повторного использования. В первом же случае (локальный ресурс) доступ к изображению будет ограничен, так как предполагается, что изображение жению будет ограничен, так как предполагается, что изображение будет использоваться только элементом управления. Рекомендуется использовать второй вариант выбора ресурса.

# 10. ЭЛЕМЕНТ УПРАВЛЕНИЯ TextBox
__Элемент управления TextBox__ обычно используется для ввода, отображения и редактирования текста. Текст, отображаемый в элементе управления, содержится в свойстве Text. __Элемент управления TextBox__ обладает дополнительными функциональными возможностями, отсутствующими у элемента управления Label, включая ввод и редактирование текста, использование многострочного режима и маскировку знаков пароля. __Для редактирования текста__ используются стандартные возможности: выделение части или всего текста, удаление отдельного символа или выделенной части (Del, BackSpace), перемещение по тексту (стрелки, Home, End и др.). С помощью клавиш Ctrl-C и Ctrl-Х из текстового поля можно копировать и соответственно вырезать выделенный текст в буфер обмена. Комбинация CtrI-V позволяет вставить из буфера обмена текст в поле TextBox. Перечислим ряд свойств элемента управления TextBox, отображаемых в окне «Свойства».
- AcceptsReturn — получает или задает значение, указывающее, какое действие необходимо выполнить в многострочном элементе управления TextBox при нажатии клавиши ENTER — создать новую строку текста или ничего не делать, активируя кнопку стандартного действия формы.
- HideSelection - получает или задает значение, показывающее, остается ли выбранный текст в поле выбранным, когда фокус ввода на форме переходит с данного элемента управления на другой.
- MaxLength — получает или задает максимальное число знаков, которое разрешается вводить или вставлять в элемент управления TextBox.
