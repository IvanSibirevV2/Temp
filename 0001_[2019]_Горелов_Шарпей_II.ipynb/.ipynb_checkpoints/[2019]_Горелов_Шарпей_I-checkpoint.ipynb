# ПРЕДИСЛОВИЕ
Разработка приложений с графическим интерфейсом, кроме написания кода, включает в себя также проектирование и разработку интерфейса пользователя. Под интерфейсом пользователя понимаются окна приложения, их элементы управления и организация взаимодействия с пользователем. Качество интерфейса, его эргономика, художественное оформление и т.д. определяют качество и привлекательность приложения. Вспомним, сколько раз менялся интерфейс таких интернет-ресурсов, как портал государственных услуг (https://www.mos.ru/) или портал Финансового университета (www.fa.ru). Разработчики этих порталов предпринимают много усилий для того, чтобы сделать свой продукт удобным. __Наиболее эффективные__ и развитые подходы к разработке Windows-приложений предлагает платформа Framework .NET и ее среда разработки Visual Studio. В этой платформе реализованы две технологии разработки интерфейса пользователя: Windows Forms и Windows Presentation Foundation (WPF). WPF является новой технологией, основанной на XML. Старая традиционная технология практически не требует знания языка XML, поэтому она проще новой. __В данном учебнике__ излагаются основы разработки Windows-приложений по технологии Windows Forms. Эта технология представлена типами пространства имен System.Windows.Forms, к которым относятся элементы управления, компоненты, диалоговые окна и другие типы. __Среда Visual Studio__ позволяет проектировать окна приложения в режиме визуальной разработки, при этом во многих случаях программисту не требуется разрабатывать код для корректировки интерфейса. Программист направляет свои усилия лишь на разработку обработчиков событий. При этом разработчик всё равно должен знать назначение, свойства и поведение элементов интерфейса и использовать наиболее подходящие элементы для решения задачи. Количество элементов управления и компонентов превышает 60. Подробно изучить их в одном учебнике не представляется возможным. В учебнике излагаются лишь наиболее часто используемые свойства и методы этих элементов, что в большинстве случаев не препятствует созданию приложения с развитым эргономичным интерфейсом. Все типовые применения элементов управления и компонентов проиллюстрированы примерами на языке С#.

# 1. ВЗАИМОДЕЙСТВИЕ ПОЛЬЗОВАТЕЛЯ С ПРОГРАММОЙ.
## 1.1. Типы программ.
Существуют два типа прикладных программ (ПП), различающихся способом взаимодействия с Windows и свойствами создаваемого окна. Это консольные приложения и приложения с графическим интерфейсом. Для консольных приложений операционная система Windows создает окно, являющееся моделью окна MS DOS с текстовым режимом. Консоль состоит из окна и клавиатуры. Манипулятор «мышь» не поддерживается. __С появлением Windows__ основными приложениями стали приложения с графическим интерфейсом. Такие приложения будем называть Windows-приложениями. __Консольная программа инициирует__ взаимодействие с пользователем с помощью запросов на выполнение того или иного метода. На рис. 1.1 показана схема взаимодействия на примере ввода с клавиатуры с помощью метода System.Console.ReadLine. Консольная программа путем выполнения методов консольного ввода (запросов) определяет моменты времени и последовательность взаимодействия программы с пользователем. __Однако Windows-приложения__ работают не так. Пользователь сам выбирает, какой элемент управления, размещенный на окне, он хочет использовать. Взаимодействие программы с пользователем берет на себя Windows. Прикладная программа теперь не запрашивает с помощью функций ввод данных, вместо этого она ждет, когда Windows известит программу о том, что пользователь выполнил какое-то действие. И к этим действиям относится не только ввод данных, но и огромное количество других действий, например выборка пункта меню. Этот процесс проиллюстрирован на рис. 1.2.

Рис. 1.1. Взаимодействие консольной программы с пользователем, управляемое запросами (функциями)

Рис. 1.2. Взаимодействие Windows-программы с пользователем, управляемое событиями

Программа ожидает сообщения, посылаемого ей операционной системой Windows. Сообщение передается в программу посредством вызова специальной функции, которая также выполняется системой Windows. После получения сообщения программа, как правило, выполняет запрограммированные действия. Взаимодействие программы и операционной системы, базирующееся на сообщениях, определяет схему построения всех Windows-программ. __В Windows существует__ множество различных типов сообщений. Например, каждый раз при щелчке мыши в пределах окна программы ей будет направлено сообщение о нажатии кнопки мыши. Клавиатурное сообщение приходит, если нажата клавиша на клавиатуре в то время, когда окно программы имеет фокус ввода (т. е. ввод с клавиатуры направляется окну). Когда программа начинает работать, сообщения к ней приходят неупорядоченным образом, т. е. программа никогда «не знает», какое сообщение будет следующим. Таким образом, Windows-программы напоминают программы, работающие по прерываниям.

# 1.2. Цикл обработки сообщений.
Как отмечалось ранее, Windows взаимодействует с программой, посылая ей сообщения. При этом существует системная очередь сообщений — одна на всю систему и по одной очереди на каждый поток. На рис. 1.3 показана схема обработки событий, реализованная в Windows. __Системная очередь__, накапливая и перераспределяя сообщения по очередям потоков, решает задачи синхронизации. Все приложения Windows должны запускать цикл обработки сообщений. __В программах__ на языке С++ цикл реализует разработчик в функции WinMain. В С#-программах такой цикл создается автоматически с помощью метода Application.Run. Этот метод в качестве параметра получает ссылку на главное окно программы. Цикл обработки сообщений выбирает сообщения из очереди сообщений приложения и направляет их обратно в Windows, которая затем вызывает оконную функцию программы, передавая ей указанные сообщения в качестве параметров. В программах на языке С++ оконная функция чаще всего имеет имя WndProc. Описанный способ взаимодействия может показаться излишне сложным, однако одной из причин реализации такой схемы является то, что программа должна возвращать управление планировщику Windows для распределения времени центрального процессора и его оптимальной загрузки.

Такой метод называется обработчиком события. Созданный каркас обработчика события будет иметь следующий вид:

    private void button1_Click(object sender, EventArgs e){}

-------

Как мы видим, имя обработчика состоит из двух частей: имени элемента управления (компонента) и имени события. По двойному щелчку элемента управления среда создаёт пустой обработчик для так называемого события по умолчанию. Для большинства элементов управления, как, например, для Button, событием по умолчанию является событие Click. __Параметры обработчика__ для отдельных событий, например Click, практически не используются. Однако для других событий эти параметры несут полезную информацию. Первый параметр описывает элемент управления (компонент), возбудивший событие, а второй — само событие. __Реализуем обработчик button1_Click__. Исходным значением для конвертирования является строка из textBox1, которая должна содержать вещественное число, следовательно, эту строку необходимо преобразовать в число. В качестве текущего курса для упрощения кода возьмем 50 рублей за 1 доллар. Вычисленное и преобразованное в строку число будем записывать в textBox2.

Приведем код обработчика button1_Click:

    float curs = 50.0f;

    private void button1_Click(object sender, EventArgs e)
    {
        float rubli, dollars;
        if (!float.TryParse(textBox1.Text, out rubli))
        {
            MessageBox.Show("неправильно введена сумма в рублях!");
            textBox1.Focus();
            return;
        }
    dollars = rubli / curs;
    textBox2.Text = dollars.ToString();
}

Создадим сборку и запустим ее на выполнение. Введем сумму 800 рублей и нажмем кнопку «Конвертировать». В результате возникнет событие Click, и обработчик button1_Click получит управление. После выполнения обработчика окно программы будет выглядеть как на рисунке 2.8.

Рис. 2.8. Окно приложения

Подробнее о добавлении обработчиков событий. Не вдаваясь в подробности, отметим, что созданию обработчика события предшествует этап подписки на это событие. Для того чтобы подписаться на требуемое событие какого-либо элемента управления и создать обработчик события, в общем случае необходимо: __во-первых__, узнать, какое имя имеет событие; __во-вторых__, выяснить, какой делегат представляет событие, и по делегату определить формат заголовка обработчика события; __в-третьих__, подписаться на событие, то есть добавить обработчик к событию; __в-четвёртых__, создать пустой обработчик события с требуемыми параметрами; __в-пятых__, разработать тело обработчика события. __Перечисленные выше шаги__ соответствуют модели событийного управления, реализованного в Windows. Выполнение их вручную — достаточно сложный для начинающего разработчика процесс, однако среда Visual Studio первые четыре шага выполняет автоматически в ответ на Ваши действия мышкой. По двойному щелчку элемента управления среда разработки подпишется на событие по умолчанию и предоставит Вам пустой обработчик для его наполнения кодом.     textBox1.TextChanged += button1_Click;

# 2.2. Подробнее о делегатах и событиях
Если для управления событиями в приложениях с графическим интерфейсом Вам достаточно приведённой выше информации, то данный раздел можно пропустить. Ниже речь пойдёт об углублённом описании понятий делегат и событие применительно к Windows-приложениям.
# ДЕЛЕГАТЫ
В событийной модели используются групповые (многоадресатные) делегаты. Делегат — это класс, содержащий данные о том, какую сигнатуру должен иметь метод, указываемый в экземпляре этого класса. В экземпляре можно определить только тот метод, у которого тип возвращаемого значения (в случае событийной модели — void) и список параметров совпадают с соответствующими элементами объявления делегата. Информация, описывающая заголовок метода, называется сигнатурой. В экземпляре делегата хранится ссылка на конкретный метод, поэтому вызов на выполнение экземпляра равносилен вызову метода. Общая форма объявления группового делегата:
    
    delegate void имя_делегата (список_параметров);

Пусть групповой делегат MyDlg объявлен следующим образом: delegate void MyDig(int a, bool b); тогда в экземпляре этого делегата в качестве параметра можно указать имя следующего метода:

    void Method (int sum, bool fl){/*...*/}

Страница 32

---------

Это допустимо, так как сигнатура метода Method соответствует делегату MyDIg. Экземпляр делегата для метода Method создается так же, как и любой объект, например

    MyDig Handler = new MyDIg (Method);

В качестве параметра конструктора указывается вызываемый метод. Поскольку этот метод будет вызывать система, то такой метод называется методом обратного вызова. Как было указано ранее, существует более простой способ создания экземпляра делегата, когда указывается только имя метода, например:

    MyDIg Handler = Method;

# СОБЫТИЯ
Событийная модель в CLR строится на основе групповых делегатов. В событийной модели применяется схема «издатель-подписчик». В этой модели в различных источниках используются термины: издатель, подписчик, уведомление, подписка, регистрация, открепление и др. В данном учебнике для упрощения материала, по возможности, будут использоваться понятия из языка С#. __На уровне Windows__ аппаратные события возникают от клавиатуры, мышки и таймера. В классах различных элементов управления присутствуют методы, создающие программные события, например «потеря элементом фокуса ввода». В языке С# событие представлено членом класса, который объявляется с помощью ключевого слова event. Синтаксис наиболее распространенной формы объявления события имеет следующий вид:

    event имя делегата имя_события;

Из этого формата следует, что событие объявляется на основе созданного ранее группового делегата. Приведем пример объявления события Click класса Button:

    public event EventHandler Click;
    
Страница 33

------

Из примера следует, что в элементе управления типа Button определено событие Click, которому может быть присвоен экземпляр делегата System.EventHandler с соответствующим по сигнатуре обработчиком события. Эта операция называется подпиской на событие. Например, по двойному щелчку элемента button1 среда Visual Studio сгенерирует (в Form1.Designer.cs) следующую подписку:
    
    button1.Click += new System.EventHandler(button1_Click);

Также в Form1.cs будет сгенерирован пустой обработчик button1_Click. Сигнатура этого обработчика будет соответствовать делегату System.EventHandler. В каждом элементе управления определено множество событий. В визуальном режиме (в окне «Свойства») разработчик может выбрать событие, возникновение которого необходимо обрабатывать, и двойным щелчком или путем выборки старого обработчика выполнить подписку. Иногда разработчику требуется выполнить подписку на событие в том месте кода, где ему это необходимо. В этом случае он может использовать упрощенный синтаксис, например, для вышеприведенного примера такая подписка будет выглядеть следующим образом:

    button1.Click += button1_Click;

Страница 34

------------

Если обработчик события имеет маленький размер, то его тело можно перенести в лямбда-выражение, например:

    button1.Click += (sender, args) => Close();

В этом случае анонимный обработчик события будет создан автоматически. Составные операции «+=» и «-=» позволяют добавлять ссылки на методы в экземпляр группового делегата и, соответственно, удалять их оттуда. Вызов события на выполнение приводит к последовательному вызову всех методов, ссылки на которые хранит экземпляр группового делегата. Вызов экземпляра также приводит к выполнению методов списка. Таким образом, член event — это дополнительный слой программного обеспечения. Такие методы называются обработчиками событий. __В классах__, представляющих элементы управления, определение событий и их вызов на выполнение осуществляется в методах, имя которых начинается с On. Так, определение события `Click` и вызов его обработчиков выполняется методом `OnClick`, а события `Enter` — методом `OnEnter`. Методы On... вызываются либо программно, либо при возникновении вложенного события более низкого уровня. Если переопределить или заместить метод On..., добавив свою обработку события, которое он вызывает, то необходимость в подписке на это событие отпадает. Приведем пример. Пусть необходимо выдать сообщение в случае щелчка формы Form1, не подписываясь на событие Click:

    protected override void OnClick(EventArgs e)
    {
        MessageBox.Show("Для формы не определена обработка события Click!*);
        base.OnClick(e);
    }

Есть одно ограничение: все методы On... защищены (Protected) и вызвать их можно только из производного класса, которым, например, является класс Form1. Однако необходимость использования методов On... возникает редко. __И последнее__. Если Вы будете реализовывать свою событийную модель в приложении, то можно отказаться от членов event и использовать в классе только экземпляры групповых делегатов. Причина, по которой для классов был введен член event, для читателя неочевидна. Разные авторы часто указывают на повышение безопасности. В частности, есть мнение, что дело заключается в том, что экземпляр группового делегата (как и многие объекты) имеет методы, с помощью которых пользователь может повлиять на чужие ссылки на методы. Событие же не является объектом и поэтому не имеет каких-либо методов. Удалим из разрабатываемого нами приложения лишние подписки и методы, созданные в качестве примеров. Рассмотрим файловый состав и код приложения.

# 3. ОБЩИЕ СВЕДЕНИЯ ОБ ЭЛЕМЕНТАХ УПРАВЛЕНИЯ.
 Элемент управления - это отдельный элемент пользовательского интерфейса, отображающий данные и/или принимающий ввод данных. В программе элементы управления являются объектами, обеспечивающими взаимодействие пользователя с программой. Каждый элемент управления представлен классом. __В среде__ разработки Visual Studio существует несколько шаблонов, по которым создаются проекты приложений с графическим интерфейсом. Для каждого шаблона используются свои элементы управления. Наиболее распространенным шаблоном для разработки Windows-приложений является Windows Forms. Альтернативным и более новым вариантом является шаблон WPF со своими элементами управления. Также свои элементы управления имеют шаблоны для ASP.NET и ADO. NET. Кроме того, система программирования С# предоставляет разработчикам средства для создания пользовательских элементов управления. __Все элементы__ проекта Windows Forms представлены в Visual Studio в окне «Панель элементов», которое можно отобразить с помощью пункта меню «Вид». Элементы представлены классами пространства имен System.Windows.Forms. Насчитывается более шестидесяти элементов, которые делятся на пять групп (Visual Studio 2015): __Во-первых__, Стандартные элементы управления (21 элемент); __Во-вторых__, Контейнеры (6); __В-третьих__, Меню (5); __В четвёртых__, Данные (5); __В пятых__, Компоненты (14); __В Шестых__, Печать (5); __В Седьмых__, Диалоговые окна (5); __В Восьмых__, Взаимодействие WPF (1).

Почти все элементы управления наследуют класс Component, поэтому они являются компонентами. Класс Component предоставляет базовую реализацию интерфейса IComponent и делает возможным совместное использование объектов разными приложениями. При этом многие элементы прямо или косвенно наследуют класс Control, который создает соответствующее изображение на форме. Элементы используются для создания пользовательского интерфейса, поэтому эти элементы называют элементами управления. Класс Control реализует базовые функциональные возможности для всех элементов управления. Рис. 3.1. Форма и панель компонентов.

Некоторые элементы управления, которые наследуют класс Control, отображаются на панели компонентов, но не отображаются на форме. В форме отображаются только коллекции этих контейнеров. __Кроме элементов__ управления, пространство имен System.Windows.Forms предоставляет классы, которые не являются производными от класса Control, но также обеспечивают визуальные функции для приложений Windows. __Ряд элементов__ управления являются контейнерами, предназначенными для хранения других элементов управления. К таким контейнерам относятся Panel, GroupBox и др. Главным контейнером элементов управления является форма, представленная классом Form. Не отображаемые элементы, то есть компоненты, в коллекцию формы не добавляются. __B Visual Studio__ все элементы пользовательского интерфейса перечислены в окне «Панель элементов». __Первый параметр__ обработчика для большинства событий один и тот же: он описывает элемент, для которого сгенерировано событие. # 5. ВНЕДРЕНИЕ ИЗОБРАЖЕНИЯ В СБОРКУ.

Многие элементы управления, имеющие поверхность, например Button, имеют свойства, позволяющие либо использовать изображение в качестве фона, либо нарисовать его поверх фона. К таким свойствам, соответственно, относятся BackgroundImage и Image. Изображение первоначально может храниться в графическом файле. Допускается множество графических типов, например jpg, png. Среда Visual Studio оформляет изображение в виде так называемого ресурса. Не вдаваясь в подробности, можно сказать, что ресурс — это данные, внедряемые в сборку. После внедрения изображения в сборку исходный графический файл станет ненужным. Когда разработчик определяет файл с изображением для элемента управления, среда Visual Studio предлагает выбрать, в какое место поместить ресурс-изображение (точнее, с каким resx-файлом ассоциировать файл ресурса) (см. рис. 5.1). Вы можете выбрать либо «Локальный ресурс», либо «Файл ресурсов проекта». Изображение в обоих случаях в виде ресурса будет внедрено в сборку, однако в последнем случае в проекте будет создана папка Resources и в неё будет скопирован файл с изображением. Этот файл всегда будет доступен разработчику для повторного использования. В первом же случае (локальный ресурс) доступ к изображению будет ограничен, так как предполагается, что изображение жению будет ограничен, так как предполагается, что изображение будет использоваться только элементом управления. Рекомендуется использовать второй вариант выбора ресурса.

# 10. ЭЛЕМЕНТ УПРАВЛЕНИЯ TextBox.
Элемент управления TextBox обычно используется для ввода, отображения и редактирования текста. Текст, отображаемый в элементе управления, содержится в свойстве Text. Элемент управления TextBox обладает дополнительными функциональными возможностями, отсутствующими у элемента управления Label, включая ввод и редактирование текста, использование многострочного режима и маскировку знаков пароля. Для редактирования текста используются стандартные возможности: выделение части или всего текста, удаление отдельного символа или выделенной части (Del, Backspace), перемещение по тексту (стрелки, Home, End и др.). С помощью клавиш Ctrl-C и Ctrl-Х из текстового поля можно копировать и, соответственно, вырезать выделенный текст в буфер обмена. Комбинация Ctrl-V позволяет вставить из буфера обмена текст в поле TextBox. Перечислим ряд свойств элемента управления TextBox, отображаемых в окне «Свойства».
- AcceptsReturn — получает или задает значение, указывающее, какое действие необходимо выполнить в многострочном элементе управления TextBox при нажатии клавиши ENTER — создать новую строку текста или ничего не делать, активируя кнопку стандартного действия формы.
- HideSelection - получает или задает значение, показывающее, остается ли выбранный текст в поле выбранным, когда фокус ввода на форме переходит с данного элемента управления на другой.
- MaxLength — получает или задает максимальное число знаков, которое разрешается вводить или вставлять в элемент управления TextBox.

В элементе управления TextBox можно вводить или отображать текст только в одном формате, например, нельзя часть текста вывести курсивом. Поскольку строк в многострочном TextBox может быть много и их длина может быть достаточно большой, то возникает необходимость использования полос прокрутки, которые можно определить свойством ScrollBars. Значения этого свойства: во-первых, None — полосы прокрутки не отображаются; во-вторых, Horizontal — отображаются только горизонтальные полосы прокрутки; в-третьих, Vertical — отображаются только вертикальные полосы прокрутки; в-четвёртых, Both — отображаются и горизонтальные, и вертикальные полосы прокрутки. В однострочном поле, если длина текста превышает размер доступного пространства, отображается лишь часть текста. В многострочном поле (MultiLine = true) эту стандартную настройку можно изменить, присваивая соответствующие значения свойствам WordWrap и ScrollBars. WordWrap показывает, переносятся ли автоматически в начало следующей строки слова текста по достижении границы многострочного текстового поля. Когда элемент управления `MaskedTextBox` отображается во время выполнения, маска в нем представляется в виде последовательности символов приглашения и дополнительных литералов. __Например__, в маске `(000)-000-0000` символы '0' являются символами приглашения, а символы '(', ')' и '-' — дополнительными литералами. Литералы всегда занимают статическую позицию в маске, и пользователь не может перемещать или удалять их. Символы приглашения доступны для изменения и представляют обязательный или необязательный входной символ. __Десятичный разделитель__ (.), разделитель тысяч (,), времени (:), даты (/) и символы валют ($\$$) по умолчанию отображаются в соответствии с языком и региональными параметрами, установленными для приложения. Например, для маски `00/00/0000` правильным форматом даты для России будет `дд.мм.гггг`, а для США — `mm/dd/yyyy`. __Можно принудительно__ отображать другие символы приглашения, используя свойство `FormatProvider`. Пользователи могут перемещаться по маске, используя клавиши со стрелками влево и вправо или курсор мыши, а ввод пробела позволяет пропустить необязательные позиции в маске.

Путем обработки этого события можно реализовать собственную логику обработки ошибок. __Литералы при вводе__ допустимого символа автоматически пропускаются. Для перемещения к следующей или предыдущей позиции можно использовать клавиши со стрелками, `Пробел` (`Space`) или `Backspace`. Также для перемещения вправо можно использовать символ приглашения.


Свойство `MaskFull` позволяет определить, введены ли пользователем все необходимые символы. Это свойство можно использовать в обработчике события `MaskInputRejected`, чтобы определить, были ли вводимые пользователем данные отклонены из-за того, что в маске не осталось позиций для ввода. __Чтобы определить__, были ли введены только необходимые элементы, используйте свойство `MaskCompleted`. __Свойство__ `Text` всегда будет извлекать введенные пользователем данные, отформатированные согласно маске и значению свойства `TextMaskFormat`, которое получает или задает значение, определяющее, включаются ли в свойство Text только литералы, только символы приглашения или символы обеих категорий. Если символы приглашения не включаются, они преобразуются в пробелы в форматированной строке. __Использование масок__ не всегда гарантирует, что вводимые пользователем данные окажутся допустимым значением для заданного типа. В следующем коде показана маска: `maskedTextBox1.Mask = "00/00/0000";` __Эта маска требует__, чтобы пользователь ввел восемь цифр, но она не может проверить, что пользователь ввел значения месяца, даты и года в допустимом диапазоне. Для этой маски одинаково приемлемы строки "20.12.2020" и "70.70.0000".

С помощью свойства ValidatingType можно задать, какому типу должно соответствовать вводимое значение. Для даты и/или времени свойству `ValidatingType` необходимо присвоить объект `System.Type` типа `DateTime`. Пример: `maskedTextBox1.ValidatingType = typeof(DateTime);` __Если свойству__ `ValidatingType` присвоено значение, то после завершения синтаксического анализа текстового поля `MaskedTextBox` (с помощью метода `Parse`) возникает событие `TypeValidationCompleted`. Это событие происходит, когда пользователь переключает фокус из элемента управления `MaskedTextBox` на другой элемент. Обработчик этого события позволяет узнать, как завершилась проверка значения. 

__Обработчику события TypeValidationCompleted__ передается параметр `e` типа `TypeValidationEventArgs`. Приведем наиболее значимые события этого параметра:
- __IsValidInput__ — получает значение, указывающее, было ли успешно выполнено преобразование входной строки в тип, заданный свойством `ValidatingType`.
- __ReturnValue__ - получает объект (`object`), являющийся результатом преобразования отформатированной ВХОДной строки. Значение свойства `ReturnValue` параметра `TypeValidationEventArgs` можно использовать после успен ной проверки поля на соответствие заданному типу.
- __Cancel__ — возвращает или задает значение, показывающе, следует ли отменить данное событие. Значение `true` отменяет событие и удерживает фокус в элементе управления `MaskedTextBox`. По умолчанию `e.Cancel = false`.

# 12. ГРУППИРОВАНИЕ ЭЛЕМЕНТОВ.
Размещая элементы управления на форме, разработчик добавляет их в контейнер этой формы. Однако форма может содержать элементы управления, которые также являются контейнерами. К числу таких контейнеров относятся элементы управления `GroupBox` (рамка) и `Panel` (панель). Существуют и другие контейнеры.

# 12.1. Элемент управления GroupBox.
Элемент управления GroupBox отображается в виде рамки с необязательным заголовком. В рамку помещают элементы управления, которые образуют группу. Как правило, с помощью `GroupBox` создают группу переключателей `RadioButton`. Иногда с эргономической целью рамку рисуют вокруг других элементов управления. На рис. 12.1 показано несколько элементов управления, помещенных в рамку. __Особенность контейнера__ `GroupBox` состоит в том, что действия, выполняемые с контейнером, распространяются на всё его содержимое. Например, для того чтобы сделать все элементы управления контейнера неактивными, достаточно сделать неактивным их контейнер, например: `groupBox1.Enabled = false;` __Чаще всего__ в программе никаких действий с элементами управления `GroupBox` не выполняется. __Объект__ `GroupBox` не предназначен для отображения полос прокрутки. Если необходим схожий элемент управления с полосой прокрутки, используйте элемент управления Panel.

# 12.2. Элемент управления Panel.
 Элемент управления `Panel` (панель) представляет собой рамку без заголовка. Кроме того, элемент управления `Panel` по умолчанию отображается без границ. Это позволяет сделать из панели невидимую кнопку. Свойство `BorderStyle` отображает стандартную или трехмерную границу панели. __Поскольку элемент__ управления `Panel` выводится из класса `ScrollableControl`, то с помощью свойства `AutoScroll` в элементе управления `Panel` можно включить полосы прокрутки. Если свойство `AutoScroll` имеет значение `true`, все элементы управления, расположенные в панели, можно прокручивать с помощью присутствующих полос прокрутки. __В остальном многое__, что было сказано выше об элементе управления `GroupBox`, справедливо и для `Panel`.

Элемент управления `CheckBox` имеет множество свойств, методов и событий. Большинство из них совпадают с аналогичными для `Button`. Есть и свои уникальные члены. Свойство __Appearance__ определяет, отображается ли элемент управления `CheckBox` в качестве обычного элемента управления `CheckBox` (по умолчанию) или как кнопка `Button`. Для того чтобы установить или прочитать состояние флажка с двумя состояниями, используется свойство `Checked`, которое может иметь значение либо `true`, либо `false`: `if ( checkBox1.Checked ) {/*...*/}`. __Для флажка с тремя состояниями__ используется свойство `CheckState`. Это свойство может принимать одно из следующих значений: Checked — элемент отмечен; Unchecked — элемент не отмечен; Indeterminate — элемент не доступен. __Если для свойства__ `ThreeState` задано значение `true`, свойство `Checked` будет возвращать значение `true` как для отмеченного флажком, так и для неопределенного состояния. На практике в основном используются __два подхода к работе с флажками_. __Во-первых__. Пользователь сначала устанавливает все флажки, а затем нажимает какую-то кнопку и в её обработчике осуществляется чтение состояний флажков. __Во-вторых__: для фиксации смены состояния флажка для него создаётся обработчик события CheckedChanged, которое является событием по умолчанию. Обработчик получает управление, если состояние флажка изменилось. Пример обработчика:

# 13.2. Элемент управления `RadioButton`
Элемент управления `RadioButton` называется переключателем. На форме, как правило, используются два или более переключателей, образующих группу (см. рис. 13.3). Каждый переключатель группы представляет взаимоисключающий вариант выбора. Переключатель может находиться в одном из двух состояний: выключен (Checked = false) или включен (Checked = true).
- [x] Переключатель 1
- [ ] Переключатель 2
- [ ] Переключатель 3

Рис. 13.3. Пример группы переключателей.

В группу должны быть объединены зависимые переключатели. Чтобы создать на форме группу или несколько групп переключателей, поместите каждую группу в собственный контейнер, такой как `GroupBox` или `Panel`. Все элементы управления `RadioButton`, не входящие в указанные (и другие) контейнеры, составляют группу контейнера формы. __На рис. 13.4__ показаны две группы переключателей, находящиеся в контейнерах `GroupBox`. В группе пользователь может выбрать и тем самым включить только один переключатель. Рекомендуется на этапе визуальной разработки в окне свойств один из переключателей группы сделать включенным по умолчанию. При выборе пользователем какого-либо выключенного переключателя группы он включается (появляется точка в кружке), а ранее включенный переключатель этой группы автоматически отключается. Элементы управления `RadioButton` и `CheckBox` имеют аналогичные функции, они позволяют пользователю выбрать вариант или отменить выбор. Разница заключается в том, что несколько элементов управления `CheckBox` можно выбрать одновременно, а элемент `RadioButton` — только один в группе.

Рис. 13.4. Две группы переключателей, объединённые контейнерами.

Чтобы получить или задать состояние элемента RadioButton, используйте свойство Checked. Основным событием переключателя, как и флажка, является CheckedChanged. Техника работы с переключателями остается практически такой же, как и с флажками. Но объем работы снижается, так как после определения установленного переключателя, остальные переключатели группы можно не анализировать. Пример. В приведенном ниже фрагменте программы осуществляется поиск первого включенного переключателя контейнера groupBox1. Предполагается, что группа, кроме флажков, может включать и другие элементы управления, поэтому выполняется анализ типа элемента управления.

# 14. РЕГУЛЯТОРЫ.
К регуляторам, описываемым ниже, относятся элементы управления `DomainUpDown` и `NumericUpDown`.
# 14.1. Элемент управления DomainUpDown.
Элемент управления `DomainUpDown` позволяет пользователям просматривать список текстовых строк и выбирать в нем отдельные строки. Этот элемент выглядит как сочетание текстового поля и пары кнопок для перемещения вверх и вниз по списку (см. рис. 14.1). Иногда элемент управления `DomainUpDown` называют регулятором.

Рис. 14.1. Элемент управления `DomainUpDown`

По сравнению с элементом управления `ComboBox`, достоинством `DomainUpDown` является его сверхкомпактное представление. Но в этом заключается и недостаток. Рекомендуется перед использованием списка отсортировать его в алфавитном порядке. Это облегчит пользователю поиск нужной строки. Для выполнения сортировки необходимо свойству `Sorted` присвоить значение `true`. Это свойство доступно в окне «Свойства». __Другими полезными свойствами__, доступными в окне «Свойства», являются `Text`, `Items`, `ReadOnly` и `Wrap`.
- Свойство __Text__ — получает или задает текст, отображаемый в текстовом поле регулятора. Этого свойства достаточно для решения большинства задач. 
- Свойство __Items__ содержит список строк, которые перебираются в элементе управления.

Пользователь может вводить в элемент управления свой текст, если свойство `ReadOnly` имеет значение `false`. Значение `true` разрешает пользователю нажимать лишь кнопки «стрелка вверх» и «стрелка вниз», благодаря чему выбор возможен лишь из числа предопределенных значений. __Если для свойства__ `Wrap` задано значение `true`, список закольцовывается, то есть при прокрутке после последней строки списка выводится первая и наоборот. __В коде программы__ разработчик может использовать следующие свойства и методы: __Свойство__ `SelectedIndex` — получает или задает значение индекса для выделенного элемента. Если элемент не выбран, индекс равен -1. __Свойство__ `SelectedItem` — получает или задает текстовое представление выделенного элемента, основываясь на значении индекса элемента, выделенного в коллекции. Если элемент не выбран, `SelectedItem` возвращает `null`. __Если пользователю__ разрешен ввод текста в поле, то нужно понимать, что для введенного вручную текста `SelectedIndex=-1`, `SelectedItem=null`, a `Text !=""`. Такие условия позволяют отличить выбранную строку от введенной.

# 14.2. Элемент управления NumericUpDown.
Элемент управления `NumericUpDown` представляет собой сочетание текстового поля и пары кнопок со стрелками для выбора значения пользователем. В текстовом поле отображается отдельное числовое значение, имеющее тип `decimal` (см. рис. 14.3).

Рис. 14.3. Элемент управления NumericUpDown.

B отличие от DomainUpDown, элемент управления NumericUpDown не содержит списка: его очередное значение вычисляется путем прибавления инкремента к текущему значению. Пользователь может увеличивать и уменьшать число, нажимая кнопки со стрелками «вверх» и «вниз» или клавиши со стрелками «вверх» и «вниз», а также вводить число в поле. Чтобы задать допустимый интервал значений элемента управления, присвойте значения свойствам Minimum и $[Maximum](Махimum). При нажатии клавиши со стрелкой «вверх» значение увеличивается до максимума; при нажатии клавиши со стрелкой «вниз» число уменьшается до минимума. Одним из вариантов применения элемента управления NumericUpDown является его использование в качестве регулятора громкости звука, поэтому иногда этот элемент управления называют регулятором.

Страница 119

----------------

Пользователь может ввести число, если только свойство ReadOnly имеет значение false. При снятии фокуса с элемента управления любые введенные данные проверяются на соответствие с минимальным и максимальным числовыми значениями и, если необходимо, сбрасываются до этих значений. Свойство Value определяет начальное значение и возвращает введенное или выбранное пользователем значение. Свойство Increment определяет шаг изменения значения, когда пользователь нажмет кнопку «вверх» или кнопку «вниз». Значения всех числовых свойств могут быть дробными числами, так как эти свойства имеют тип decimal. Внимание. При вводе дробного числа в окне свойств в качестве разделителя необходимо использовать запятую, а не точку, как принято в языке С#. Внимание. При вводе дробного числа в окне свойств в качестве разделителя необходимо использовать запятую, а не точку, как принято в языке С#. Чтобы задать число десятичных разрядов, которые будут отображаться в поле, присвойте свойству `DecimalPlaces` значение, равное этому числу. Чтобы в десятичных числах отображался разделитель групп разрядов, присвойте свойству ThousandsSeparator значение true. Чтобы значения отображались в шестнадцатеричном формате, присвойте свойству Hexadecimal значение true.

Как заполнять списочные элементы управления данными. Создать коллекцию элементов можно либо в режиме визуальной разработки, либо вручную в режиме кода. Метод Add может добавить любой объект в конец коллекции списочного элемента управления. Метод AddRange добавляет в конец коллекции массив объектов. Для вставки объекта в требуемое место списка следует использовать метод Insert. Существует еще один способ добавления элементов: использование внешнего списка, например списка List<T>. Этот способ реализуется посредством привязки элемента управления к источнику данных с помощью свойства DataSource. Этому вопросу посвящен отдельный раздел данного учебника. Свойства DisplayMember и ValueMember. Выше было сказано, что свойство DisplayMember используется для определения, какое свойство объекта должно отображаться в элементе. Пример. Используем свойство DisplayMember для отображения свойства Name. listBox2.DisplayMember = "Name"; Свойство DisplayMember должно получать в качестве значения строку с именем открытого свойства, а не с именем открытого поля. # 17.2. Элемент управления ListBox.
Элемент управления `ListBox` отображается в виде списка строк, из которых пользователь может выбрать нужную щелчком мыши. Список ограничивает выбор пользователя. Во всех списках выбранная строка выделяется цветом и остается такой, пока не будет выбрана другая строка. Часто элемент управления `ListBox` используют вместо многострочного элемента `TextBox` для вывода информации. И для этого есть причины: во-первых, не нужно заботиться о разделении строк и, во-вторых, нужную строку можно выделить и получить ее для обработки. Пользователь не может свернуть список или как-то уменьшить его в размерах, поэтому простой список, как правило, занимает достаточно много места на форме. По этой причине на формах простые списки можно встретить нечасто. Элемент управления `ListBox` содержит свойство `MultiColumn`, с помощью которого можно отобразить список элементов в виде нескольких вертикальных столбцов. Такой способ позволяет отобразить больше элементов списка и устраняет необходимость его прокрутки для поиска нужной позиции. Не путайте эту возможность с отображением разных свойств объекта в разных столбцах. Также `ListBox` может поддерживать выбор нескольких элементов, для чего предусмотрено свойство `SelectionMode`. Это свойство принимает одно из следующих значений одноименного перечисления:
- __None__ — выделение элементов запрещено;
- __One__ — разрешено выделение только одного элемента;
- __MultiSimple__ - разрешено выделение нескольких элементов:
- __MultiExtended__ — разрешено выделение нескольких элементов, и пользователь может производить выделение с помощью клавиши `SHIFT`, клавиши `CTRL` и клавиш со стрелками.

# 17.3. Элемент управления ComboBox.
Элемент управления ComboBox представляет собой комбинацию текстового поля и простого списка. Причем по умолчанию (свойство __DropDownStyle__ равно __DropDown__) отображается только текстовое поле. Такое компактное представление комбинированного списка делает его популярным элементом управления. Благодаря таким особенностям комбинированный список часто называют полем с выпадающим списком. По щелчку кнопки со стрелкой список раскрывается. - __DropDown__ — указывает, что список отображается при нажатии кнопки «стрелка вниз», а текстовая часть является редактируемой (см. рис. 17.6, слева). - __DropDownList__ - указывает, что список отображается при нажатии кнопки «стрелка вниз», а текстовая часть является нередактируемой (см. рис. 17.6, в центре). Текстовое поле такого элемента управления отображается серым цветом.
- __Simple__ — указывает, что список всегда раскрыт, а текстовая часть является редактируемой (см. рис. 17.6, справа). Такой список занимает почти столько же места, как и список `ListBox`. Пользователь может ввести любое значение и не ограничен выбором из существующих значений списка.

Примечание: когда Вы помещаете раскрытый список на форму, не забудьте растянуть его по вертикали, иначе списка Вы не увидите.

# 17.4. Элемент управления CheckedListBox.
Список с флажками (CheckedListBox) — это постоянно раскрытый список ListBox, где слева от каждой строки отображается флажок (см. рис. 17.1, 17.2). Заголовок класса CheckedListBox: `public class CheckedListBox : ListBox`. Как следует из заголовка, элемент управления `CheckedListBox` является расширением элемента управления `ListBox`. Он выполняет практически все функции списка, а кроме того, в нем может отображаться галочка рядом со строкой списка. Другое различие между этими элементами управления заключается в том, что в списках с отмеченными элементами можно выделить только один элемент или не выделить ни одного. Установить или сбросить флажок можно только у выбранного (выделенного) элемента. Чтобы установить или сбросить флажок, можно использовать также двойной щелчок мышки на невыделенном элементе. Элемент с установленным флажком называется отмеченным. Текущее выделение элемента снимается, когда выделяется другой элемент, однако это не изменяет состояние флажка. В списке могут одновременно находиться несколько элементов с установленными флажками и только один с выделенным текстом. На рис. 17.1 показаны два элемента с отмеченными флажками и один элемент выбранный. Если выбранный элемент с неотмеченным флажком щелкнуть, то флажок будет установлен (см. рис. 17.2). Если флажок у выбранного элемента был ранее установлен, то он будет сброшен. Поскольку элемент управления `CheckedListBox` наследует члены класса `ListBox`, то для работы с коллекцией всех элементов `(CheckedListBox.ObjectCollection)` используются те же средства, что и для элемента управления `ListBox`: `Items`, `Text`, `SelectedIndex`, `SelectedItem`, `Sorted`, `UseTabStops` и др.

Каждый элемент списка CheckedListBox, независимо от того, выделен он или нет, может иметь одно из трех состояний, определяемых флажком. Эти значения являются членами перечисления CheckState: во-первых, Unchecked — флажок снят (не отмечен); во-вторых, Checked — флажок установлен (отмечен); в-третьих, Indeterminate — флажок находится в неопределенном состоянии. Такой флажок обычно затенён (недоступен). По умолчанию это значение не используется. Установить это значение можно только в коде, например, с помощью метода SetItemCheckState. # 17.5. Изменение объектов коллекции.
Если коллекция списочного элемента управления состоит из значимых типов данных, то их изменение сразу же отображается на экране. Однако, если коллекцией списочного элемента управления являются объекты, а точнее, ссылки на объекты, то при изменении какого-либо объекта на экране продолжает отображаться старое значение и перерисовать его не удается, даже если использовать методы `Refresh` или `Update`. Рассмотрим это явление на примере. Пусть у нас имеется простейший класс: 
    
    class MyClass

        public int P1 { get; set; }
        public int P2 { get; set; }
    }

В обработчике `Form1_Load` мы заполняем элемент управления `listBox` ссылками на объекты класса `MyClass`:

    private void Form1_Load(object sender, EventArgs e)

        listBox1.Items.Add(new MyClass { P1 = 2, P2 = 92 });
        listBox1.Items.Add(new MyClass { P1 = 3, P2 = 93 3);
        listBox1.Items.Add(new MyClass { P1 = 4, P2 = 94 });
        listBox1.DisplayMember = "P1";

    }

# 18. ПРИВЯЗКА К ИСТОЧНИКУ ДАННЫХ.
### 18.1. Общие сведения
Рассматривая ранее списочные элементы управления, мы не раз отмечали, что эти элементы могут заполняться данными из внешнего источника. Это заполнение можно выполнить вручную, с помощью циклов. Например, пусть в нашем распоряжении имеется список list типа List<string> и мы хотим отобразить строки этого списка в элементе управления comboBox1, для чего разрабатываем следующий метод:

    //--Создать список ComboBox--
    private void CreateComboBoxList<string> list)
    {
        comboBox1.Items.Clear();
        foreach (var g in list)
            if (IcomboBox1.Items.Contains(g))
                comboBox1.Items.Add(g);
    }

Метод `CreateComboBox` получает в качестве аргумента список и записывает его уникальные строки в элемент управления `comboBox1`. Метод `Contains` определяет, входит ли указанный элемент в данную коллекцию, при этом метод учитывает регистр, то есть строки «жасмин» и «Жасмин» — это разные строки.

Страница 175

----------

__Примечание__. В данном учебнике в примерах часто используются имена элементов управления, генерируемые по умолчанию средой разработки Visual Studio. Эти имена образуются от типа элемента управления с добавлением порядкового номера элемента в форме. Например, элементы управления `label1` и `label2` — это первый и, соответственно, второй объекты типа `Label`. Аналогично: `textBox1` и `textBox2` — это первый и, соответственно, второй объекты типа `TextBox`. То же правило применяется и к другим элементам управления. Используя такие имена объектов, мы будем считать, что читатель понимает, от какого класса они созданы. Следует отметить, что если какой-либо один или оба списка будут изменяться в процессе выполнения программы, то разработчик должен решить задачу синхронизации списков: изменится один список, требуется изменить и другой. Поэтому использование внешних данных для разработчика представляет определенную трудность. Но, к счастью, система программирования для решения этой задачи предоставляет специальный механизм, позволяющий выполнить так называемую привязку элемента управления к источнику данных, используя специальные свойства элемента управления. При этом разработчик освобождается от решения вопросов загрузки данных и их синхронизации. Под привязкой визуального элемента управления к источнику данных понимают процесс, при котором элемент управления выбирает данные из указанного для него источника. При этом исходная коллекция элементов перед привязкой очищается.

Механизм привязки устанавливает взаимосвязь между элементом управления и данными его источника, при этом изменение данных в источнике приводит к изменению их отображения в элементе управления. Это в идеале. На практике для многих поточников надо использовать специальный метод, который приводит к обновлению привязки и отображению элемента управления. В зависимости от того, каким является элемент управления — простым (не содержащим коллекции элементов) или списочным (содержащим коллекции элементов), — используются разные механизмы привязки. К простым элементам управления относятся `TextBox`, `Label` и др., к списочным — `ListBox`, `ComboBox`, `DataGridView` и др. Для привязки простых элементов управления используют простую привязку к одиночному значению, а списочных — сложную привязку к списку значений. Для простой привязки используется класс Binding, а для сложной — BindingSource. Эти классы исследуют источник данных, причем во внимание принимаются только открытые свойства. Открытые поля игнорируются. Поскольку для реализации простой привязки рекомендуется использовать класс BindingSource, рассмотрим сначала сложную привязку как наиболее востребованную. Особое место в технологии привязок занимает окно «Источники данных». В этом окне отображаются источники данных в проекте, такие как базы данных, веб-службы и объекты. Можно создавать элементы управления с привязкой к данным, перетаскивая элементы из окна «Источники данных» на форму. Можно также осуществить привязку существующих элементов управления к данным путем перетаскивания объектов из окна источников данных на существующие элементы управления.

# 18.2. Сложная привязка ОСНОВНЫЕ СВЕДЕНИЯ.
Привязка называется «сложной» не потому, что она является сложной для использования разработчиком (как раз здесь всё просто), а потому что источник данных является сложной, многострочной структурой данных. Чаще всего — это список (массив) объектов, которые заданным образом необходимо отобразить в многострочном элементе управления. К элементам управления, поддерживающим сложную привязку к источнику, относятся `ListBox`, `ComboBox`, `DataGridView`, `DataView`, `BindingSource` и др. Для реализации привязки элементов управления к источнику данных эти элементы предоставляют свойство DataSource, которое возвращает или задает источник данных для элемента управления (компонента).

### Примечание
Поскольку независимо от списочного элемента управления для привязки используются одни и те же свойства, все примеры будут приведены для элемента управления `listBox` типа `ListBox`. В качестве источника данных будем рассматривать список `list` типа `List<MyClass>`, содержащий объекты следующего класса:

    class MyClass
    {
        public int P1 ( get; set;
        public int P2 ( get; set; )
    }

Используются два способа сложной привязки элемента управления: непосредственно к источнику данных и через посредника.

__Пример__. Привязать `listBox1` непосредственно к списку объектов `list` типа `List<T>`. `listBox1.DataSource = list;` При использовании привязки элемента управления к источнику данных возникает ряд проблем. Главная из них состоит в следующем: как известно, источник данных в процессе работы программы чаще всего претерпевает изменения: добавляются новые объекты, какие-то удаляются, свойства объектов получают новые значения. Если источник изменился, то необходимо, чтобы привязанный к этому источнику элемент управления был об этом извещен. Тогда элемент управления перерисует отображение объектов на экране. Так должно быть. Однако на практике многие источники данных — массивы, списки `List<T>` и др. — никаких уведомлений не посылают. К числу источников данных, которые уведомляют привязанные элементы управления, относятся `BindingList`, `BindingSource` и некоторые другие.

Страница 178
    
--------

Кроме того, источники данных могут иметь разную природу и отличаться друг от друга. Не со всеми источниками данных одинаково успешно может взаимодействовать элемент управления. Поэтому для решения этой проблемы был разработан компонент `BindingSource`, который можно использовать в качестве универсального связующего средства между элементами управления в форме и источником данных (см. рис. 18.1). То есть `BindingSource` является посредником, который берет на себя функции детального исследования источника данных и уведомления привязанного к нему элемента управления о необходимости перерисовать список. Класс `BindingSource` иногда называют соединителем.

- Элемент управления <-> BindingSource <-> Источник данных

Рис. 18.1. Использование посредника в сложной привязке.

При использовании соединителя элемент управления с помощью свойства `DataSource` привязывается не к источнику данных, а к компоненту `BindingSource`, который для элемента управления становится источником данных. Компонент `BindingSource` (объект `bindingSource1`), в свою очередь, привязывается к источнику данных с помощью своего свойства `DataSource`.

__Пример__. Использование соединителя для привязки элемента управления `listBox1` к списку `list`.

    bindingSource1.DataSource = list.
    listBox1.DataSource = bindingSource1;

# Источники данных.
В качестве источников данных для сложной привязки можно использовать массивы `Агау`, коллекции `ArrayList`, `List<T>` и `BindingList<T>`, таблицы `DataTable` и представления `DataView ADO.NET` и многие другие. Можно не перечислять все источники, а ограничиться следующим комментарием: при сложной привязке обычно требуется, чтобы объект поддерживал интерфейс `IList` или `IListSource`. Так как эти интерфейсы определяют список объектов, организованных в коллекцию, то сложная привязка данных также называется привязкой на основе списка.

---------- 

Если при привязке используется компонент `BindingSource`, можно выполнить привязку к объекту, поддерживающему также интерфейс `IEnumerable`. Таким объектом, например, является последовательность, возвращаемая `LINQ`-запросом. (Присвятой LINQ To Objects.) Компонент `BindingSource` копирует все элементы `IEnumerable` в отдельный список `List`, предназначенный исключительно для привязки. Все коллекции и массивы (а также класс `CollectionBase`) реализуют интерфейс `IList`. Эти классы представляют собой индексированные списки объектов типа `Object`. Такие элементы управления называют объектами на основе списков. Обратите внимание, какое количество интерфейсов наследует класс `List<T>`:

    public class List<T> : IList
        , IList<T>
        , ICollection, ICollection<T>
        , IEnumerable, IEnumerable<T>

Как видите, если есть универсальный интерфейс, то должен быть и неуниверсальный. Класс, реализующий интерфейс `IListSource`, обеспечивает привязку на основе списка для объектов, не являющихся списками. Метод `GetList` объекта `IListSource` используется, чтобы возвратить связываемый список из объекта, не наследующего от списка `IList`. Объект `IListSource` используется классом `DataSet`. Компонент `BindingSource` расширяет перечень доступных для привязки источников данных. К примеру, такие элементы управления, как `DataGridView` и `ComboBox`, не поддерживают привязку непосредственно к источникам данных `IEnumerable`, однако такую привязку можно осуществить посредством компонента `BindingSource`. В этом случае компонент `BindingSource` преобрадует источник данных в интерфейс `IList`. Можно создать источники данных из баз данных, служб или объектов. Можно перетащить элементы из окна «Источник данных» в рабочую область конструирования проекта для создания элемента управления с привязкой данных к свойствам выбранного объекта. Список значений в качестве источника данных можно создать с помощью запроса Linq.

# Отображаемое и возвращаемое свойства источника.
У элементов управления ListBox и ComboBox есть несколько полезных свойств, часто используемых совместно с привязкой:
- __DisplayMember__ — определяет свойство источника данных, значение которого будет отображаться в элементе управления.
- __ValueMember__ - возвращает или задает свойство источника данных, значение которого извлекается из текущего объекта.
- __SelectedValue__ — возвращает или задает значение свойства, определенного свойством ValueMember.

У сетки DateGridView и компонента BindingSource перечисленных свойств нет, но есть свойство DataMember, которое указывает подсписок или таблицу источника данных. __Если объекты__ источника данных имеют несколько открытых свойств, то требуется уточнение — какое свойство источника следует отображать на экране в списке ListBox или ComboBox, а какое свойство будет возвращать значения. Это уточнение делается с помощью таких свойств списка, как DisplayMember и (соответственно) ValueMember. Значением каждого из этих свойств является имя свойства источника, представленное в виде строки, например, для свойств класса, приведённого выше, конкретизация свойств может выглядеть следующим образом:

    listBox1.DisplayMember = "P1";
    listBox1.ValueMember = "P2";

Следует отметить, значение свойства P2 выбирается из объекта, соответствующего выбранной, то есть текущей строке списочного элемента управления.

# Члены класса BindingSource и область их применения.
Приведем описание некоторых членов класса `BindingSource`. В описании используется термин «базовый источник данных». Этим источником является список объектов, к которому привязан `BindingSource`.

### Свойства класса `BindingSource`:
- __this[индекс]__ - индексатор. Возвращает или задает элемент базового списка по указанному индексу.
- __List__ — получает базовый список, к которому привязан соединитель.
- __Count__ — получает общее число элементов в базовом списке.
- __Current__ — получает текущий элемент в базовом списке.
- __DataMember__ — возвращает или задает список в источнике данных, к которому в настоящий момент привязан соединитель.
- __DataSource__ — возвращает или задает источник данных, к которому привязан соединитель.
- __Position__ - возвращает или задает индекс текущего элемента в базовом списке.

Свойством по умолчанию в классе `BindingSource` является свойство `DataSource`. Методы `BindingSource` для работы со списком источника данных:
- __Add__ — добавляет существующий элемент в базовый список.
- __AddNew__ — создает новый объект с помощью конструктора по умолчанию и добавляет его в базовый список.
- __ApplySort__ — сортирует источник данных, используя задавные описания сортировки.
- __CancelEdit__ - отменяет текущую операцию редактирования
- __Clear__ — удаляет все элементы из базового списка, представ ленного свойством `List`, и устанавливает свойство `Count` в значение О.
- __СоруТо__ — копирует содержимое списка, представленно свойством List, в указанный массив, начиная с указанного индекса. Этот метод служит для объединения элементов из нескольких неточников в один массив.
- __EndEdit__ - применяет незавершенные изменения к базовому неточнику данных.
- __Fin__ — возвращает индекс элемента в базовом списке, имеющего заданное имя свойства и значение.
- __IndexOf__ - осуществляет поиск указанного объекта и возвращаст индекс первого вхождения в пределах списка.
- __Insert__ — вставляет элемент в список по указанному индексу.
- __Remove__ — удаляет заданный элемент из базового списка.

- __RemoveAt__ — удаляет элемент по указанному индексу из базового списка.
- __RemoveCurrent__ - удаляет текущий элемент из базового списка.

- __ResetItem__ — вызывает в элементе управления, привязанном к компоненту `BindingSource`, повторное считывание элемента списка по указанному индексу и обновление его отображаемого значения.
- __ResumeBinding__ - вызывает привязку данных.

В классе `BindingSource` есть несколько событий. Событием по умолчанию является `CurrentChanged`. Если элемент управления `listBox1` привязан к соединителю `bindingSource1`, который в свою очередь привязан к списку `list` типа `List<T>` или `BindingList<T>`, то выполняются следующие зависимости:

    bindingSource1.DataSource эквивалентно list
    bindingSource1.List эквивалентно list
    bindingSource1.DataSource эквивалентно bindingSource1.List
    listBox1.Items[1] эквивалентно list[1]
    listBox1.Items не эквивалентно list

То есть соединитель в своих свойствах List и DataSource хранит ссылку непосредственно на источник данных. Списковый элемент управления также хранит ссылки на объекты источника.

# Изменение источника данных.
Ряд из перечисленных выше методов обеспечивает косвенное обращение компонента BindingSource к источнику данных, например, метод RemoveAt компонента перенаправляется источнику данных. Поэтому источники данных, например List<T>, должны содержать соответствующие методы, что гарантируется унаследованными интерфейсами.
    
__Пример__. В следующих операторах осуществляется добавление нового объекта в базовый список и удаление из этого списка объекта с индексом 0. Метод `Add` соединителя выполняет как минимум две функции: вызывает метод `List<MyClass>.Add` базового списка, используя ссылку на него в свойстве `BindingSource.List`, и уведомляет элемент управления об изменении базового списка: элемент управления будет перерисован.

    bindingSource1.Add(new MyClass { P1 = 555, P2 = 555 });
    bindingSource1.RemoveAt(0);

Таким образом, соединитель BindingSource упрощает взаимодействие элемента управления с источником данных за счет создания уведомлений об изменении списка источника, который (список) не создает уведомлений. Для решения задач изменения списка источника данных можно использовать методы не соединителя, а самого списка.

__Пример__. В следующих строках демонстрируется работа с базовым списком, минуя соединитель. Осуществляется добавление нового объекта и удаление объекта, находящегося на нулевой позиции.

Чтобы отобразить отсортированные данные в элементе управления, необходимо отсортировать источник данных.

Вместо этого для привязки используйте элемент управления `ComboBox` или `ListBox` или выполняйте заполнение списка `CheckedListBox` вручную.

# ПЕРЕРИСОВКА ЭЛЕМЕНТА УПРАВЛЕНИЯ.
Если для изменения базового списка используются методы источника данных, то будет ли перерисован привязанный элемент управления или нет, зависит от источника данных. В приведенных примерах методы источника List<MyClass> не уведомляют соединитель о выполненных изменениях. Поэтому типична ситуация, когда список был изменен, а элемент управления продолжает отображать старые значения. Если для изменения источника данных используются методы класса BindingSource, то функцию уведомления элемента управления берет на себя соединитель — элемент будет перерисован. Другой хороший способ — вместо класса List<T> использовать обобщенный класс BindingList<T>, методы которого уведомляют привязанный элемент управления об изменениях своего списка. Класс находится в пространстве имен System.ComponentModel. Этот класс представляет собой универсальную версию интерфейса IBindingList, который расширяет интерфейс IList, добавляя свойства, методы и события, необходимые для двусторонней привязки данных ($[BindingSource](BindingSourse) — `BindingList). __Класс__ BindingList<T>реализует функциональность, близкую к классуList<T>, поэтому замена одного имени класса на другой $[осуществляется](осушествляется) без каких-либо проблем. Правда, класс BindingList<T>не имеет своего метода сортировкиSort, поэтому вместо него нужно использовать метод расширения OrderBy. Кроме того, вы можете отказаться от соединителя и привязывать элемент управления непосредственно к коллекции BindingList<T>, так как класс BindingList<T>, по своей сути, является некоторой комбинацией классов BindingSourceиList<T>. В частности, класс Binding и List<T>имеет методResetBindings, что позволяет обновлять привязку. __До сих пор речь шла об__ использовании методов класса BindingSourceили классовList<T>иBindingList<T>`. А что происходит, если изменяется отдельное значение какого-нибудь открытого свойства объекта источника обычным присваиванием? Сразу скажем: как правило, при изменении свойства никаких уведомлений соединителю не посылается. Рассмотрим пример, в котором разными способами извлекается ссылка на объект и одно из свойств этого объекта получает новое значение.

Повторное считывание уведомляет элемент управления о необходимости обновить отображение (одной или всех строк).

# РЕЖИМ ВИЗУАЛЬНОЙ ПРИВЯЗКИ.
Компонент `BindingSource` находится в окне элементов, поэтому его можно перетащить на форму. При этом автоматически создастся объект, а на панели компонентов отобразится имя этого объекта. Далее Вы можете выполнить привязку соединителя к источнику данных, используя окно свойств. На этом этапе Вы можете привязаться, как правило, только к типу объектов источника данных. Такую привязку рекомендуется выполнять для элемента управления `DataGridView`, в котором открытые свойства объектов источника отображаются в виде столбцов с заголовками, соответствующими именам свойств.

# ИСПОЛЬЗОВАНИЕ yield ДЛЯ ФОРМИРОВАНИЯ ИСТОЧНИКА ДАННЫХ.
Класс `BindingSource` позволяет привязать данные, реализующие в упрощённом виде лишь один интерфейс `IEnumerable`. Поэтому возможна, например, привязка, в которой очередное значение возвращается из перечислителя с помощью ключевого слова `yield`.

    yield return ‹expression>;
    yield break;

Выражение `expression` вычисляется и возвращается в виде значения объекту перечислителя; выражение `expression` должно неявно преобразовываться в тип результата итератора. На внутреннем уровне (за «кулисами») при обработке ключевого слова `yield` компилятор динамически генерирует в рамках соответствующего класса определение вложенного класса. В этом классе создаются метод `MoveNext` и свойство `Current`. (`Метод Reset` не генерируется.) Оператор `yield` не обязан быть последним в блоке цикла. Если он находится в середине, то выполнение блока продолжается со следующего оператора после `yield`.

# 18.3. Простая привязка ОСНОВНЫЕ СВЕДЕНИЯ.
Для привязки простого элемента управления к источнику данных следует использовать свойство DataBindings этого элемента управления. (Речь идет об элементах управления, у которых отсутствует свойство DataSource.) Свойство DataBindings используется для доступа к коллекции ControlBindingsCollection привязок свойств элемента управления к источникам данных. Привязка является объектом типа Binding, который определяет параметры привязки, то есть коллекция ControlBindingsCollection состоит из этих объектов. В отличие от сложной привязки списочных элементов управления, простая привязка является двусторонней: изменяется источник — изменяется и элемент управления, изменяется элемент управления — изменяется и источник. Однако в обратную сторону — от элемента управления к источнику — привязка срабатывает не всегда, как хотелось бы. Предлагаю читателю поэкспериментировать со свойствами Text двух элементов управления TextBox, привязывая один к другому. Достоинство привязки заключается в том, что где бы в программе ни изменялось свойство элемента управления или его источник данных, разработчик, если он установил привязку, освобождается от контроля за этими изменениями. Замечание. Следует понимать, что речь идет об источниках, свойства которых оповещают привязку об изменениях своих значений. К таким источникам относятся, например, свойства элементов управления. Если Вы привяжете элемент управления к свойству, не поддерживающему привязку, то это свойство не будет оповещать элемент управления о своих изменениях. Свойство DataBindings определяет не один элемент, а коллекцию привязок. Причиной этого является то, что у одного элемента управления есть несколько свойств, каждое из которых может быть привязано к своему источнику данных. Например, для свойства Text элемента label1 может использоваться одна привязка в коллекции, а для свойства ForeColor этого элемента — другая. Однако одно свойство элемента управления не может быть привязано к нескольким источникам данных. Для добавления объекта Binding в коллекцию используется метод Add коллекции, который имеет 7 перегрузок. Приведем первые четыре версии метода Add, используемые в данном учебнике:

# СОБЫТИЯ КЛАССА Binding.
С объектом Binding связано три события:
- __Parse__ — происходит при изменении значения элемента управления с привязкой к данным.
- __Format__ - вызывается при передаче данных из источника данных в элемент управления. В обработчике события можно выполнить преобразованиие неформатированных данных из источника данных в форматированные данные для вывода на экран.
- __BindingComplete__ - происходит, когда свойство FormattingEnabled получает значение true, и когда завершается операция привязки, например, при передаче данных из элемента управления в источник данных или наоборот.

При передаче данных из элемента управления в источник данных сначала вызывается событие Parse, которое позволяет отменить форматирование отображаемого значения, а затем — событие Format, которое переформатирует данные для вывода на экран. Обработчики этих событий в качестве параметра не получают ссылку на передаваемое из источника значение.

__Пример__. В приведённом ниже коде показано, как подписаться на события для преобразования и форматирования выводимого значения, и приведены примеры соответствующих обработчиков, осуществляющих преобразование значения типа `decimal`. (Полный текст можно найти в справочной системе языка C#.)

        ConvertEventHandler(CurrencyStringToDecimal);
        bind.Format += new ConvertEventHandler(DecimalToCurrencyString);

# 19. ГЛАВНОЕ И КОНТЕКСТНОЕ МЕНЮ.
К числу часто используемых элементов управления, но вместе с тем достаточно простых с точки зрения их обычного применения, относятся меню и панель инструментов. Меню может быть главным, состоящим из ниспадающих (раскрывающихся) пунктов, и контекстным, вызываемым по правой кнопке мыши. Рассмотрим эти элементы управления.

Пользователи достаточно часто сталкиваются с меню в различных приложениях, например в `Microsoft Office`. Среда `Visual Studio` позволяет создать меню, аналогичное `Microsoft Word`. По сути, меню — это одна или несколько коллекций элементов, чаще всего представленных кнопками. Каждая кнопка называется командой, а элемент, объединяющий кнопки, — родительским меню. Элемент меню иногда будем также называть пунктом меню. Основное достоинство меню заключается в его компактном отображении: команд не видно, пока Вы их не раскроете, щелкнув родительское меню. __Примечание__. Двойной щелчок элемента меню в форме `Visual Studio` приводит к созданию обработчика события `Click` с соответствующим именем, например: `void команда1ToolStripMenuItem_Click(object sender, EventArgs e)`. __Как видите__, первая часть имени обработчика — это имя элемента меню. Рекомендуется это имя сократить. __Все команды меню должны иметь свои обработчики__. Для ниспадающего меню редко создают обработчик события `Click`, так как при выборке такого элемента выполняется стандартная операция: осуществляется отображение вложенных в контейнер элементов. Однако, если обработчик определен для меню, он будет вызван. __Во время выполнения__ программы команда меню может быть выполнена с помощью щелчка левой кнопкой мыши, с помощью горячих клавиш (сочетания клавиш) и клавиши быстрого вызова. Удобство использования меню можно также повысить с помощью добавления флагов и значков.

# 20.2. Практика использования компонента `ToolStrip`
Перетащите на форму из окна элементов компонент `ToolStrip`. Под заголовком формы появится макет элемента со значком по умолчанию. Щелкнув этот значок, Вы получите первую кнопку панели `ToolStrip`. # 20.4. Дополнительные возможности `ToolStrip`
Использование элементов управления `ToolStrip` имеет богатый опыт разработки, включая активацию на месте и редактирование, пользовательский макет и совместное использование горизонтального или вертикального пространства в указанном `ToolStripContainer`. Класс `ToolStrip` предоставляет множество элементов, обеспечивающих управление рисованием, вводом с помощью мыши и клавиатуры, а также функции перетаскивания. Для получения больших возможностей настройки и контроля стиля оформления и макета для всех элементов управления `ToolStrip` в `Windows Form` можно использовать класс `ToolStripRenderer` совместно с классом `ToolStripManager`. Можно объединить элементы управления `ToolStrip` в рамках заданного объекта `ToolStripContainer` и произвести слияние элементов управления `ToolStrip` друг с другом. Для приложений многодокументного интерфейса (`MDI`) используйте `ToolStripPanel` вместо объекта `ToolStripContainer`. Обычно объект `ToolStrip` не применяется в последовательности перехола, за исключением случаев, когда он имеет абсолютное положение, а не закреплен или находится в `ToolStripPanel`.

Поскольку класс ToolStrip является общим не только для меню и панели инструментов, но и для такого элемента управления, как строка состояния, рассмотрим коротко этот элемент. Строка состояния представлена классом StatusStrip и во многом напоминает панель инструментов. По умолчанию строка состояния минает панель инструментов. По умолчанию строка состояния прижимается к нижней границе формы (элемента управления) и показывает текущие данные, например: текущее время, номер текущей страницы и т.д. __Во время разработки__ для элемента управления StatusStrip доступны следующие типы: `ToolStripStatusLabel`; `ToolStripDropDownButton`; `ToolStripSplitButton`; `ToolStripProgressBar`. __Как правило__, элемент управления `StatusStrip` состоит из объектов `ToolStripStatusLabel`, каждый из которых отображает текст, значок или и то, и другое. __Поскольку строка__ состояния в основном используется для отображения данных, для её элементов, как правило, обработчики событий не создаются.

# 22.1. Общие сведения.
Таймер для приложений `Windows Forms` представлен компонентом `Timer`, находящимся в пространстве имен `System.Windows.Forms`. Данный таймер разработан для однопоточной среды `Windows`. Класс `Timer` достаточно значим, так как ряд задач не может быть решен без использования таймера. Основное назначение таймера — вызов события `Tick` через определенные пользователем интервалы времени и вызов обработчика этого события.  В режиме визуального проектирования, после перетаскивания таймера на форму, создается обработчик событий `Тіск`. Как правило, это выполняется с помощью двойного щелчка таймера на панели компонентов. После истечения временного интервала таймер не прекращает свою работу, а начинает очередной отсчет интервала. В разделе 23.4 приведен пример программы, в которой создается окно, плавно появляющееся и плавно исчезающее с экрана. Плавность достигается путем постепенного изменения прозрачности формы в обработчике событий таймера. __Класс `Timer`__ имеет небольшое количество важных членов, к числу которых относятся следующие:
- __Свойство `Interval`__ — определяет временной интервал между событиями `Tick` в миллисекундах.
- __Свойство `Enabled`__ - запускает (`true`) или останавливает (`false`) таймер.
- __Метод `Start`__ — запускает таймер. Начинает вызывать событие `Tick`, задавая для свойства `Enabled` значение `true`.
- __Метод `Stop`__ — прекращает вызывать событие `Тіск`, задавая для свойства `Enabled` значение `false`.

__Работа таймера__ заключается в генерации событий `Tick` через период времени `Interval`. Таймер может быть запущен, если `Interval` > 0. Таймер можно остановить, обнулив значение свойства `Interval`.

# 23.1. Что такое `SDI-` и `MDI-`интерфейсы.
В зависимости от характера поставленной задачи разработчик может создать многооконное приложение с однодокументным (`SDI`) или многодокументным интерфейсом (`MDI`). Приложения с многооконным интерфейсом `MDI` позволяют отображать несколько документов сразу, при этом каждый документ отображается в отдельном окне, как в приложении `MS Word`. Приложения `SDI` позволяют одновременно открывать только одно окно документа. Понятие «документ» для интерфейса `SDI` является условным, так как зачастую никаких документов приложение не обрабатывает. Если не замыкаться на документах, то можно дать такое определение: в многооконном приложении с интерфейсом `SDI` все окна являются модальными; в многооконном приложении с интерфейсом `MDI` отдельные окна являются немодальными. __Модальное окно__ — это такое активное окно приложения, с которого нельзя переключиться на любое другое окно этого приложения, пока не будет закрыто текущее. __Пример__. Окно, отображаемое следующим вызовом, является модальным. `MessageBox.Show("Модальное окно");`

__Немодальное окно__ не препятствует пользователю переключаться на другие окна данного приложения, не закрывая активное. Такая модель как нельзя лучше подходит для приложений, запрограммированных для одновременной обработки нескольких документов. Приложения с интерфейсом `SDI` — это традиционные приложения, они проще приложений с интерфейсом `MDI`, поэтому мы начнем изучение взаимодействия окон с изучения интерфейса `SDI`. Независимо от вида интерфейса приложения, все окна являются объектами базового класса `Form`, имеющего богатую функциональность. Изучению этой функциональности посвящен следующий подраздел. Если одна форма создает другую, то первая является __родительской__, а вторая — __дочерней__.

# 23.2. Базовый класс Form.
__Класс `Form`__ является представлением любого окна, отображаемого в приложении. Этот класс используется для создания стандартных окон, окон инструментов, не обрамленных и перемещаемых окон. Класс `Form` также используется для создания модальных окон, например диалоговых окон. Особым видом формы является форма интерфейса `MDI`, содержащая другие формы, называемые дочерними `MDI`-формами. `MDI`-форма создается посредством задания для свойства `IsMdiContainer` значения `true`. Дочерние `MDI-формы` создаются с помощью установки свойства `MdiParent` в значение, равное ссылке на родительскую `MDI-форму`, которая будет содержать дочернюю форму. __Класс `Form`__ является наследником многих классов, включая `Control` и `Component`, и с этой точки зрения является отображаемым элементом управления. Однако в окне элементов этот класс не представлен, так как одна форма не может включать в свой контейнер другую форму. Тем не менее форма может создать другую, дочернюю форму и передать ей управление в одном из своих обработчиков событий. __С помощью свойств__, доступных в классе `Form`, имеется возможность определить внешний вид, размер, цвет и компоненты управления создаваемого окна или диалогового окна. Свойство `Text` позволяет задавать заголовок окна в строке заголовка. С помощью свойств `Size` и `DesktopLocation` определяются размер и положение окна при его отображении. Имеется возможность использовать свойства `Font` и `ForeColor` для изменения назначенного по умолчанию шрифта и цвета текста всех элементов управления, помещенных в форму. Свойство `BackColor` меняет цвет фона формы. Свойства `FormBorderStyle`, `MinimizeBox` и `MaximizeBox` позволяют управлять возможностью свертывания, развертывания или изменения размера формы во время выполнения.

При вызове этого метода родительская форма передает управление в дочернюю форму, и код, следующий за методом `ShowDialog`, не выполняется до закрытия дочерней формы. Так как `ShowDialog` не создает новых экземпляров, то его можно выполнять повторно — все старые значения формы будут видны. Немодальные окна отображаются с помощью других средств.

# 23.3. Закрытие модального окна.
Как было отмечено выше, родительская форма может отобразить дочернее модальное окно методом `ShowDialog`. При этом родительская форма блокируется до завершения работы с дочерней. Завершить свою работу дочерняя форма может, закрыв себя методом `Close` или системной кнопкой закрытия. При закрытии формы происходит несколько событий в следующей последовательности: `FormClosing` — попытка закрыть форму; `FormClosed` — после закрытия формы; `Deactivate` — после закрытия формы. Если обработчик события `FormClosing` установит `e.Cancel = true`, то форма останется открытой, `e.Cancel = false` — форма закроется, и возникнут события `FormClosed` и `Deactivate`. Если дочерняя форма была отображена с помощью метода `ShowDialog` и она закрывается методом `Close`, окно скрывается с экрана, но не удаляется. При этом метод `Dispose`, определенный во второй части класса формы (в файле `Formx.Designer.cs`), не получает управления. Если требуется очистить все используемые управляемые ресурсы, вызовите метод `Dispose` вручную. Если же закрывается главная форма, метод `Dispose(true)` класса формы автоматически получает управление.

 После вызова `Close`, выданного для закрытия текущего окна, окно исчезает с экрана, при этом программа продолжает свою работу с оператора, следующего за `Close`. То есть форма не прекращает свое выполнение! Это может повлечь серьезные проблемы. Рассмотрим следующий пример, код которого приведен ниже. Пусть форма закрывается, если некоторый файл, например `File1.txt`, существует, иначе файл создается.

    private void button1_Click(object sender, EventArgs e)
    {
        if (File.Exists("File1.txt"))
            Close();
        // Файла нет, создаем новый.
        File.Create("File1.txt");
        // Файла нет, создаем новый.
        File.Create("File1.txt");
    }

Если файл `File1.txt` существует, то форма закрывается, то есть становится невидимой, однако обработчик `button1_Click` продолжит выполнение операторов. Следующим оператором после `Close` выполнится создание нового файла `File1.txt`, что приведет к очистке существующего файла, то есть пользователь потеряет информацию, хранящуюся в файле. Следовательно, код примера является некорректным. Правильным решением является следующее:

    private void button1_Click(object sender, EventArgs e)
    {
        if (File.Exists("File1.txt"))
        {
            Close();
            return;
        }
        // Файла нет, создаем новый.
        File.Create("File 1,txt");
    }

Таким образом, после оператора `Close` рекомендуется выполнить действия, приводящие к выходу из обработчика. Это может быть или использование оператора `return`, или передача управления на конец обработчика. По той же причине не рекомендуется помещать `Close` во внутренние методы, вызываемые из обработчиков событий. Приведем пример некорректного завершения программы из внутреннего метода. Пусть обработчик `button1_Click` вызывает метод `Method1`, в котором выполняется закрытие формы.

    private void button1_Click(object sender, EventArgs e)
    {
        Method1();
        // Эти операторы будут выполнены после Close().
    }
    private void Method1() { Close(); }

В этом примере, после закрытия формы в методе `Method1`, обработчик `button1_Click` продолжит выполнение операторов, расположенных ниже вызова метода (отмечены многоточием). Рассмотрим процесс закрытия формы более подробно. Вызов метода `Close` приводит к выполнению обработчиков событий `FormClosing` и `Form1_FormClosed`, если они есть. Для главной формы выполняется метод Dispose, находящийся в файле `Designer.cs`. Далее, окно скрывается с экрана (`Visible = false`). Если форма является главной, то в очередь сообщений записывается сообщение с кодом 0 для завершения программы. Для того чтобы обработать это сообщение, Ваша программа должна выйти из того обработчика событий, в котором находится в текущий момент времени. Обработка сообщения с кодом 0 приводит к уничтожению формы (`this = null`) и передаче управления в метод `Main` для его завершения. Один из способов удаления дочерней формы — удалить ссылку на неё. Если ссылка на дочернюю форму будет удалена, то она будет уничтожена сборщиком мусора (правда, не сразу). Другой способ принудительного завершения программы основан на возможности операционной системы завершить любой процесс. Для этого после Close используем функцию `Environment.Exit(exitCode);` Метод `Environment.Exit` завершает текущий процесс и возвращает операционной системе указанный код выхода. Ресурсы этот метод не освобождает; рекомендуется предварительно их освободить методом `Dispose`.

# СОЗДАНИЕ ОКНА НЕСТАНДАРТНОГО ВИДА.
Форма и элементы управления используют регионы для определения своей конфигурации. Регион — это графическая фигура, границы которой определяются объектом класса GraphicsPath. Операционная система не отображает части окна, лежащие вне региона окна. Для установления своих границ форма и элементы управления могут использовать свойство Region, унаследованное от класса Control. Свойство Region не отображается в окне свойств. # ВСПЛЫВАЮЩЕЕ ОКНО.
Ниже показан способ плавного появления и плавного исчезновения окна. Таким окном может быть заставка с логотипом разработчика, названием приложения и т. д. Плавное появление и исчезновение окна базируется на использовании свойства `Form.Opacity`, которое определяет уровень прозрачности формы. Значение может меняться от 0.0 — совершенно прозрачная форма, до 1.0 — форма непрозрачная. Если менять прозрачность формы от 0.0 до 1.0 с некоторым шагом, то окно будет плавно проявляться на экране. __Плавное изменение__ прозрачности может быть основано на использовании компонента `Timer`, который имеет два основных свойства: `Interval` и `Enabled`. Свойство `Interval` определяет временной интервал между событиями `Tick` в миллисекундах. А свойство `Enabled` запускает (`true`) или останавливает (`false`) таймер. Работа таймера заключается в генерации событий `Tick` через интервал `Interval`. __Приведем пример__ фрагментов программы, выполняющей плавное появление и плавное исчезновение окна. Для этого создадим приложение с двумя формами — главной (`Form1`) и формой-заставкой (`Form2`). В главной форме создадим обработчик `Form1_Load`, в котором отобразим вторую форму на экране.

    private void Form1_Load(object sender, EventArgs e)
    { 
        Form2 frm2 = new Form2();
        frm2.ShowDialog();
    }

В форме-заставке, используя свойство `BackgroundImage` и свой рисунок, определите фон формы. Установите свойство формы `Opacity` в значение 0. Перетащите на форму таймер `Timer`. В окне свойств таймера установите начальные значения:

    timer1.Enabled = true;
    timer1.Interval = 25;

# ЗНАЧОК ДЛЯ ФОРМЫ.
Каждая форма может иметь свой значок, который устанавливается свойством `соп`. Это свойство задает или возвращает значение типа `Icon`. Тип `Icon` является растровым изображением определенного формата с прозрачным фоном. Значок представляет форму на панели задач, а также в заголовке формы. Свойство `Iсоп` не будет иметь никакого эффекта, если значение свойства `FormBorderStyle` определяет заголовок без значка (см. рис. 23.2). Значки содержатся в файлах с расширением `ісо`. __Один `ICO`-файл__ содержит один или несколько значков, размер и цветность каждого из которых задаётся отдельно. Например, значок для операционной системы `Windows ХР`, изготовленный по рекомендациям фирмы `Microsoft`, содержит в себе 9 отдельных изображений. Это связано с тем, что при уменьшении растрового изображения оно теряет резкость, а при увеличении — детализацию. Это особенно заметно, когда уменьшению или увеличению подвергаются значки маленького размера. 

__Как правило__, значки представляют собой растровые картинки квадратного формата стандартных размеров. В операционных системах `Microsoft Windows` наиболее употребимы значки со стороной 16, 32 и 48 пикселей. В различных средах также используются значки с размером 24, 40, 60, 72, 92, 108, 128 пикселей и другие. С ростом разрешения мониторов современные операционные системы включают поддержку всё более крупных размеров значков — до 256×256 для `Windows Vista`. __Начиная с `Windows 98/2000`__, поддерживается внедрение сжатых изображений в формате `JPEG` и `PNG`, но обычно данные значков хранятся в несжатом виде. Значки внедряются в программу в виде ресурса. Для создания значка с помощью встроенного в `Visual Studio` графического редактора необходимо щелкнуть правой кнопкой на проекте, далее выбрать: Add -> New Item... -> Icon File -> Добавить... Запустится графический редактор со своим главным меню и панелью инструментов. Особенности создания значка в этом редакторе было описано в разделе 2.

# ИСПОЛЬЗОВАНИЕ КЛАВИШ ENTER И ESC.
Обычно для завершения ввода текста с клавиатуры требуется щелкнуть клавишей мыши другой элемент управления, например кнопку Button, однако это связано с переключением руки с клавиатуры на мышь и с потерей времени на позиционирование курсора. К счастью, ввод можно завершить клавишей `Enter`, если с помощью свойства формы AcceptButton указать, какая кнопка формы будет срабатывать при нажатии этой клавиши. Срабатывание кнопки означает выполнение обработчика события этой кнопки. На рис. 23.6 показано, что при выборе свойства AcceptButton среда перечисляет для выбора все доступные на форме кнопки.

Рис. 23.6. Выбор кнопки для клавиши Enter.

Однако такой способ будет работать только в том случае, если в текущий момент на форме нет элемента управления, который перехватывает клавишу Enter и обрабатывает ее. Например, элемент управления многострочного текстового поля допускает нажатие клавиши Enter, если она выбрана для вставки нового строкового символа в элемент управления. Необходимо учитывать этот факт. Кроме клавиши Enter, с кнопкой Button можно связать и клавишу Esc. Что будет выполняться при нажатии клавиши Esc или соответствующей ей кнопки Button, решает разработчик. Например, можно очистить текстовое поле или выполнить закрытие формы. Свойство `CancelButton` может не работать, если другой элемент управления в форме перехватывает действие клавиши `Esc`. Например, если в форме открыто поле со списком `ComboBox`, нажатие клавиши `Esc` приведет к закрытию поля со списком `ComboBox`.

# КОГДА ПЕРВОЕ ОТОБРАЖАЕМОЕ ОКНО — ВСПОМОГАТЕЛЬНОЕ.
Как правило, главная форма приложения является основной формой, и в ней выполняется основная работа. Однако, если в начале работы программы требуется ввести в текстовые поля логин с паролем, то либо эти поля должны постоянно присутствовать на главной форме, либо в качестве главной формы надо использовать вспомогательное окно для ввода пароля. В последнем случае вспомогательная форма будет постоянно присутствовать на экране до конца работы программы. Ниже приведен альтернативный пример, когда главная форма, будучи не отображенной, вызывает в обработчике `Form1_Load` дочернее окно ввода пароля.

 В примере предполагается, что во второй форме по результатам ввода и анализа пароля формируется открытое поле `public bool password;`

# ПЕРЕДНИЙ И ЗАДНИЙ ПЛАНЫ.
Все элементы управления помещаются в контейнер формы, определяемый свойством `Form.Controls`. От порядка следования элементов зависит, какой из них будет на переднем плане, а какой — на заднем. Понятие «передний-задний план» имеет смысл для элементов, перекрывающих друг друга. Если элементы накладываются друг на друга, то, как правило, отображением одного из них управляют с помощью свойства `Visible`. Приведем следующий пример. Пусть для ввода в текстовое поле должен использоваться элемент управления `MonthCalendar` (календарь), который вызывается двойным щелчком текстового поля. Будем считать календарь вспомогательным элементом. По умолчанию календарь невидим (`monthCalendar1.Visible = false`), но в обработчике двойного щелчка текстового поля календарь будет отображен (`monthCalendar1.Visible = true`). Календарь занимает достаточно много места, поэтому чаще всего он перекрывает какие-либо элементы управления. Однако может случиться, что календарь отобразится на заднем плане, то есть его будут закрывать элементы управления. Возникает вопрос, как переместить календарь на передний план? Какой элемент управления будет на переднем плане, а какой на заднем, определяет последовательность добавления элементов в контейнер формы. Элемент, добавленный раньше, будет на переднем плане по сравнению с элементом, добавленным позже.

---------

__Пример__. Разместить `textBox1` на передний план.

    Controls.Add(textBox1);
    Controls.Add(button1);
Разместить `button1` на передний план.

    Controls.Add(button1);
    Controls.Add(textBox1);
Элементы добавляются в контейнер формы во второй части класса `Formx`. Однако исправлять вторую часть формы не рекомендуется. Проще воспользоваться кнопками панели инструментов 

(левая кнопка — «На передний план», правая — «На задний план»). Размещение вспомогательного элемента на переднем плане с помощью перечисленных выше приемов не всегда является хорошим решением, так как в процессе визуального проектирования формы вспомогательный элемент будет загораживать другие элементы. Для решения этой проблемы рекомендуется перемещать вспомогательный элемент на передний план не во время работы в конструкторе формы, а во время выполнения программы. Для этого используем метод `BringToFront`.

# РАСПОЛОЖЕНИЕ ФОРМЫ.
Свойство `StartPosition` возвращает или задает начальное положение окна на экране в режиме выполнения. Имеется возможность расположить форму в центре экрана или в центре ее родительской формы. Возможные значения свойства (константы перечисления `Form.StartPosition`):
- __Manual__ - положение формы формы определяется свойством `Location`.
- __CenterScreen__ — форма с заданными размерами располагается в центре текущего отображения.
- __WindowsDefaultLocation__ - форма с заданными размерами размещается в расположении, определенном по умолчанию в `Windows` (значение по умолчанию). Как правило, каждая следующая форма размещается под заголовком предыдущей со сдвигом вправо.
- __WindowsDefaultBounds__ - положение формы и ее границы определены в `Windows` по умолчанию. Форма может увеличить или уменьшить свой размер, так как по умолчанию он равен 300×300 точек.
- __CenterParent__ - форма располагается в центре родительской формы.

Иногда необходимо получить размер окна, например, для того чтобы вычислить, не выйдет ли за границу динамически отображаемый элемент управления. Другая задача, которую приходится решать, — размещение формы в требуемом месте рабочего стола. Эту задачу можно решить с помощью свойства `Location` или следующего метода формы: `public void SetDesktopLocation(int x, int y)`, где х и у — координаты левого верхнего угла формы на рабочем столе. Свойство `Location` — возвращает или задает объект `Point`, который определяет верхний левый угол формы в экранных координатах.

# РАЗМЕРЫ ФОРМЫ.
При рисовании или размещении на форме каких-либо объектов часто бывает необходимо знать размеры формы и её частей. Форма сохраняет свои линейные размеры в свойствах. Ряд этих свойств перечислен ниже:
- __ClientRectangle__ — получает прямоугольник, представляющий клиентскую область формы.
- __ClientSize__ — возвращает или задает высоту и ширину клиентской области формы.
- __Size__ — возвращает или задает высоту и ширину формы.
- __Bottom__ — возвращает расстояние между нижним краем формы и верхним краем клиентской области контейнера. Значение данного свойства равно сумме значений свойств `Тор` и `Height`.
- __Bounds__ — возвращает или задает размер и местоположение формы, включая не клиентские элементы. Местоположение определяется относительно клиентской области контейнера формы. Границы формы включают не клиентские элементы, такие как полосы прокрутки, границы, заголовки и меню. Метод `SetBoundsCore` вызывается для настройки свойства `Bounds`. Свойство `Bounds` не всегда удается изменить через метод `set`, поэтому следует переопределить метод `SetBoundsCore`, чтобы гарантировать выполнение кода в случае, если задано свойство `Bounds`.
- __Width__ - возвращает или задает ширину формы.
- __Height__ — возвращает или задает высоту формы.
- __Тор__ - возвращает или задает координату по оси `Y` верхнего края формы (в точках).
- __Left__ - возвращает или задает координату по оси `Х` левого края формы (в точках).
- __Right__ — возвращает расстояние от правого края формы до левого края контейнера. Значение свойства `Right` равно сумме значений свойств `Left` и `Width`.

Поскольку многие из перечисленных свойств унаследованы от класса `Control`, то эти свойства доступны также и элементам управления. Свойство `ClientRectangle` возвращает прямоугольник, задающий клиентскую область (рабочую область) формы. Клиентская область формы располагается между границами формы, не включая в себя неклиентские элементы, такие как границы и головки окон. В клиентскую область формы входят полоски меню (`MenuStrip`), инструментов (`ToolStrip`), состояния (`StatusStrip`), полосы прокрутки (`VScrollBar`, `HScrollBar`) и т.д. Если форма имеет подзаголовочные полоски, то начало координат будет закрыто этими полосками.

# АВТОМАТИЧЕСКОЕ МАСШТАБИРОВАНИЕ ФОРМЫ.
Приложение может быть перенесено на другой компьютер, имеющий ряд параметров, не совпадающих с параметрами компьютера, на котором было разработано данное приложение. В том случае окна приложения могут выглядеть не так, как планировал разработчик. Однако класс `Form` может автоматически изменять свои размеры и свое содержимое в соответствии либо с текущим системным шрифтом, либо с разрешением экрана (в приках на дюйм). __Если разработчик установит__ для формы `AutoScaleMode = None`, то окно не будет масштабироваться, и это приведет к тому, что на других компьютерах окно будет выглядеть либо слишком маленьким, либо слишком большим, а текстовые поля могут не вместиться (а в элементы управления. Например, если приложение разработано с использованием системного `шрифта Tahoma` размером 9 пунктов, то без коррекции окно будет выглядеть слишком маленьким на компьютере, на котором в качестве системного шрифта используется `шрифт Tahoma` размером 12 пунктов. Текстовые элементы, такие как заголовки, меню, содержимое текстовых полей и т.д., будут меньше, чем в других приложениях. __Аналогичная ситуация возникает__, когда приложение разработано для определенного экранного разрешения. Без коррекции окно приложения, особенно графическое, разработанное для одного разрешения, будет отображаться либо слишком большим, либо слишком маленьким при отображении на экране с другим разрешением. __Автоматическое масштабирование решает__ эти проблемы путем автоматического изменения размеров формы и ее дочерних элементов управления согласно относительному размеру шрифтов и разрешению экрана. Для решения задач автоматического масштабирования используются следующие свойства: `AutoSizeMode`; `AutoSize`; `AutoScaleMode`. В современных средах `Visual Studio` по умолчанию устанавливаются следующие значения перечисленных свойств: `AutoScaleMode = Font`; `AutoSize = false`; `AutoSizeMode = GrowOnly`;

### Рассмотрим эти свойства.
Свойство `AutoScaleMode` получает или задает режим автоматического масштабирования формы. Значением свойства является одна из констант одноименного перечисления:
- __None__ — автоматическое мясттабирование отключено.
- __Font__ — масштаб элементов управления относительно размеров шрифта, используемого классами, который обычно является системным шрифтом.
- __Dpi__ — масштаб элементов управления относительно разрешения дисплея. Обычно разрешение имеет значение 96 и 120 точек на дюйм.
- __Inherit__ - масштаб элементов управления в соответствии с родительским режимом масштабирования классов. Если родительский объект отсутствует, автоматическое масштабирование отключено.

Масштабирование с использованием значения `Font` может оказаться полезным, если следует растянуть или сжать элемент управления или форму в соответствии с размером шрифтов операционной системы. Это масштабирование должно использоваться, если абсолютный размер элемента управления или формы не имеет значения. В большинстве бизнес-приложений должен использоваться режим автоматического масштабирования `Font`. Масштабирование с использованием Dрі полезно, если следует задать размер элемента управления или формы относительно экрана. Например, можно использовать масштабирование в точках на дюйм элемента управления, отображающего диаграмму или другой график, чтобы он занимал определенное процентное соотношение экрана. Режим масштабирования Dpi полезен для графических приложений.

Свойство `AutoSizeMode` возвращает или задает режим, с помощью которого форма или элемент управления может автоматически изменять свои размеры. Режим определяется одноименным перечислением, имеющим следующие значения:
- __GrowAndShrink__ — размеры элемента управления увеличиваются или уменьшаются по размеру его содержимого. Размеры элемента управления невозможно изменять вручную.
- __GrowOnly__ - размеры элемента управления могут изменяться в соответствие с его содержимым, однако они не уменьшаются до значения, меньшего заданного свойством Size. Размеры формы можно изменять, но их нельзя сделать настолько малыми, что какие-либо содержащиеся в ней элементы управления становятся скрытыми.

Страница 299

-----------

По умолчанию используется значение `GrowOnly`. Установка значения `GrowAndShrink` приводит к тому же поведению, которое имеет место для элементов управления с разрешенным свойством `AutoSize`, однако не имеющим свойства `AutoSizeMode`. Свойства `MinimumSize` и `MaximumSize` выполняются, но текущее значение свойства `Size` игнорируется. Свойство `AutoSize` изменяет размеры формы во время выполнения в соответствии со значением `AutoSizeMode`. Значение true задает автоматическое изменение размеров формы; `false` — размеры формы необходимо изменять вручную. По умолчанию `AutoSize=false`. Используйте `AutoSize=true`, чтобы принудительно изменить размеры формы для вмещения всего содержимого. Пример.

    this.AutoSize = true;
    this.AutoSizeMode = AutoSizeMode.GrowAndShrink;
Форма по-разному реагирует на значения свойств `AutoSizeMode` и `AutoSize` во время визуального проектирования и во время выполнения приложения. Форма не меняет размеры автоматически в конструкторе форм `Visual Studio` при установке значений указанных свойств. Однако во время выполнения форма правильно изменяет собственные размеры, основываясь на значениях свойств `AutoSizeMode` и `AutoSize`.

Свойство `AutoSize` имеет не только форма, но и ряд элементов управления, например `Button`. Это свойство получает или задает значение, указывающее, основано ли изменение размеров элемента управления на его содержимом. Значение `true` определяет автоматическое изменение размеров элемента управления в зависимости от его содержимого; в обратном случае — значение `false`. __В отличие от формы__ элементы управления автоматически изменяют свои размеры как во время визуальной разработки, так и во время выполнения. __При использовании__ `AutoSize` свойства `MinimumSize` и `MaximumSize` формы и элементов управления выполняются, но текущее значение свойства `Size` игнорируется.

# 24. ВЗАИМОДЕЙСТВИЕ ФОРМ ПРИЛОЖЕНИЯ.
Под взаимодействием форм понимается их взаимодействие по управлению и данным. Взаимодействие по управлению предполагает создание и отображение форм в соответствии со сформированными условиями. Взаимодействие по данным — это обмен данными между формами приложения. __Разработчик может создать приложение__, состоящее из нескольких главных форм, последовательно запускаемых в методе `Main`. Однако наиболее часто разработчики реализуют приложения, состоящие из одной главной формы и нескольких дочерних. Каждая из дочерних форм также может стать родительской, если создаст новую форму.
# 24.1. Взаимодействие родительской и дочерней форм.
Особенностью взаимодействия родительской и дочерней форм является создание в родительской форме дочерней и передача ей управления. Выполнение родительской формы приостанавливается (режим `SDI`). После закрытия дочерней формы управление возвращается назад. На рис. 24.1 приведен пример, в котором форма `Form1` является родительской по отношению ко всем остальным формам, а форма `Form3` — дочерней по отношению к форме `Form1` и родительской по отношению к форме `Form4`. Начинающие программисты наделяют форму какими-то магическими особенностями, хотя с точки зрения объектно-ориентированного программирования форма — это обычный объект. Таким образом, взаимодействие форм — это взаимодействие объектов.

# ИСПОЛЬЗОВАНИЕ ОБЪЕКТНЫХ ЧЛЕНОВ.
Поскольку родительская форма создает дочернюю, то в распоряжении родительской формы, естественно, есть ссылка на дочернюю. Через эту ссылку, пока она не вышла из области видимости, родительская форма имеет доступ ко всем открытым нестатическим (объектным) членам дочерней формы. __Рассмотрим взаимодействие на примере__. Пусть в дочерней форме определены два открытых поля:

    public string title;
    public bool closeFlag;
Тогда родительская форма может до запуска дочерней передать ей значение для `title` и после закрытия дочерней формы получить значение поля `closeFlag`. Во фрагменте родительской формы, приведенном ниже, показано, как это можно сделать. В примере родительская форма закрывается, если в дочерней возникла непреодолимая ошибка.

Родительская форма:

    // Создание дочерней формы.
    Form2 form2 = new Form2();
    // До запуска form2.
    form2.title = "Финансовый университет";

    form2.ShowDialoa():

Самый простой способ решения этой задачи — передать ей эту ссылку так, как это продемонстрировано в следующем примере. Перенесем открытое поле `closeFlag` из дочерней формы в родительскую. Определим в дочерней открытое поле `form1` типа `Form1`. Родительская форма:

    public bool closeFlag;

    Form2 form2 = new Form2();
    // Конструктор получает управление.
    // Передаем ссылку на себя в form2.
    form2.form1 = this;
    form2.ShowDialog(); // Отображаем form2.

Дочерняя форма:

    public Form1 form1;
    //......
    // Использовать поле из родительской формы.
    form1.closeFlag = true;
Дочерняя форма получает доступ к открытому полю `closeFlag` родительской формы, используя ссылку `form1`. Нельзя использовать ссылку на родительскую форму в конструкторе дочерней формы. Связано это с тем, что конструктор дочерней формы получает управление при создании экземпляра формы, а ссылка передается позже. Поэтому в конструкторе дочерней формы `form1 = null`. __В предыдущем примере__ `form2` является локальной переменной обработчика события (метода), и эта переменная будет уничтожена после выхода из обработчика. Если открытые поля и свойства дочерней формы должны быть доступны в разных местах родительской формы, то рекомендуется объявление ссылки на дочернюю форму вынести на уровень полей класса, тогда дочерняя форма не будет уничтожена сборщиком мусора при выходе из обработчика события, создавшего дочернюю форму.

Например, родительская форма:

    public Form2 form2;
    // Объявление поля.
    //...........
    // Создание дочерней формы.
    form2 = new Form2();
    // Конструктор получает управление.
    //...............
Использование метода `Application.OpenForms`. Ссылку необязательно передавать в дочернюю форму, ее можно получить, используя класс `Application`. Статическое свойство `OpenForms` этого класса возвращает коллекцию открытых форм, из которой можно выбрать ссылку на требуемую форму: `Form1 form1 = (Form1)Application.OpenForms["Form1"];`

# ИСПОЛЬЗОВАНИЕ ВОЗВРАЩАЕМОГО ЗНАЧЕНИЯ.
Метод `ShowDialog`, отображающий дочернее окно, может возвратить значение, показывающее, как завершилось это окно. Результат диалога для формы возвращается или задается свойством `Form.DialogResult`. (В окне свойств `Visual Studio` это свойство не отображается.) Свойство `Form.DialogResult` может получить одно из следующих значений перечисления `DialogResult`:
- __None__ — из диалогового окна возвращается значение `Nothing`. Это означает, что модальное диалоговое окно продолжает свою работу.
- __ОК__ — из диалогового окна возвращается значение `ОК` (обычно это значение отправляется кнопкой с надписью «`ОК`»).
- __Cancel__ — из диалогового окна возвращается значение `Cancel` (обычно это значение отправляется кнопкой с надписью «`Cancel`»).
- __Abort__ — из диалогового окна возвращается значение `Abort` (обычно это значение отправляется кнопкой с надписью «`Abort`»).
- __Retry__ — из диалогового окна возвращается значение `Retry` (обычно это значение отправляется кнопкой с надписью «`Retry`»).
- __Ignore__ — из диалогового окна возвращается значение `Ignore` (обычно это значение отправляется кнопкой «`Ignore`»).
- __Yes__ — из диалогового окна возвращается значение `Yes` (обычно это значение отправляется кнопкой с надписью «`Yes`»).
- __No__ — из диалогового окна возвращается значение `No` (обычно это значение отправляется кнопкой с надписью «`No`»).

Если форма отображается в виде диалогового окна, установка свойства `Form.DialogResult` со значением из перечисления `DialogResult` (любым, кроме `None`) устанавливает значение результата диалогового окна для формы, скрывает модальное диалоговое окно и возвращает дочернюю форму вызывающей форме (родительской). __Свойство `Form.DialogResult`__ для возврата значения из диалоговой формы может быть установлено программистом в обработчике события. Это свойство нельзя устанавливать в конструкторе формы, поэтому `DialogResult` отсутствует в окне свойств формы.

Кроме разработчика, свойство `Form.DialogResult` также устанавливается автоматически кнопкой `Button`, для которой установлено свойство `Button.DialogResult`. Значение свойства `Button.DialogResult` присваивается свойству `Form.DialogResult`, поэтому свойство `Button.DialogResult` имеет такой же тип, как и `Form.DialogResult`. Как видите, `DialogResult` — это свойство формы, свойство кнопки и перечисление (не путайтесь). __Разработчик должен выбрать__, какие кнопки диалоговой формы и какое значение типа `DialogResult` они должны возвращать. Далее разработчик должен войти в окно свойств кнопки и присвоить свойству `DialogResult` выбранное из списка значение. На рис. 24.3, слева, показан пример диалогового окна с двумя кнопками. Каждая из этих кнопок является финишной, то есть закрывающей окно. На рис. 24.3, справа, показано, как для свойства `DialogResult` кнопки выбирается значение `DialogResult.Yes`.

Рис. 24.3. Пример диалогового окна и значения свойства `DialogResult`

Когда пользователь нажимает кнопку `Button`, для которой определено свойство `DialogResult` со значением, отличным от `None`, это значение присваивается свойству формы `DialogResult`. Далее управление получает обработчик события кнопки, если он есть. Если определены обработчики событий `FormClosing` и `FormClosed`, то они также получают управление. После этого модальное диалоговое окно скрывается, но не уничтожается, и управление возвращается родительской форме. __Скрытая форма может__ быть отображена снова, при этом не требуется в родительской форме создавать новый экземпляр диалогового окна. Из-за этого поведения необходимо вызвать метод формы `Dispose`, если форма уже не нужна приложению. __Если свойство `Form.DialogResult`__ не установлено или окно закрывается системной кнопкой закрытия «X» или или свойством `Visible = false`, всегда возвращается значение `DialogResult.Cancel`. Обработчик события `FormClosing` может изменить это значение или отменить закрытие формы. __Метод `ShowDialog`__ возвращает результат также в виде перечисления `DialogResult`, имеющего показанные на рис. 24.3 значения. __Если Вы собираетесь вручную установить__ для кнопки свойство `DialogResult`, то этого не надо делать в обработчике этой кнопки. Лучше воспользуйтесь обработчиком `Formx_Load`.

# 24.2. Управление формами в Main
В методе `Main` можно создать несколько независимых форм. Все формы являются главными формами приложения, поэтому каждая следующая форма создается только после закрытия предыдущей (см. рис. 24.5). Пока метод `Main` не завершится и ссылка не выйдет из зоны видимости, имя формы будет оставаться на панели задач. Чтобы убрать это имя, необходимо получить ссылку и присвоить ей значение `null`.

Рис. 24.5. Последовательное выполнение трех форм.

__Пример__. Метод `Main` запускает первую форму, а после ее завершения — вторую.

    static class Program
    {
        [STAThread]
        static void Main()
        {
            Application.EnableVisualStyles();
            Application.SetCompatibleTextRenderingDefault(false);
            Form1 frm1 = new Form1();
            Application.Run(frm1); // Сделать главной форму 1.
            frm1 = null; // Уничтожить форму 1.
            Application.Run(new Form2()); // Сделать главной форму 2.
        }
    }
Такие формы могут взаимодействовать друг с другом на уровне полей класса, содержащего метод `Main`. Если вторая форма должна запускаться при условии, формируемом первой формой, то это условие можно проанализировать в методе `Main`.

# 25. НЕМОДАЛЬНЫЕ ОКНА.
Выше было сказано, что окно называется немодальным (нережимным), если оно отображается методом `Show`, который не блокирует родительскую форму. Это означает следующее: __Во-первых__. Управление в родительской форме после создания немодальной формы немедленно передается на следующий оператор после Show. __Во-вторых__. Родительская форма продолжает получать сообщения. __В-третьих__. Родительская форма не получает никакого уведомления о том, что дочерняя форма закрылась. __В-четвертых__. С немодального окна, имеющего фокус ввода (активное окно), можно переключиться на родительское окно и на другое немодальное окно. С родительского окна можно переключиться на немодальное дочернее окно. Модальное окно, как известно, не позволяет переключаться ни на какие другие окна своего приложения. __В-пятых__. Если закрывается родительское окно, то закрываются и все его немодальные дочерние окна. Параллельное выполнение родительского и дочерних немодальных окон делает программу сложнее.

Рассмотрим, в каких случаях может потребоваться создание немодального окна. Во-первых, для реализации интерфейса `MDI`, основанного на немодальных окнах. Во-вторых, в рамках `SDI`-приложения иногда полезно отдельное окно сделать немодальным. Например, инструкцию о программе, представленную отдельной формой, можно отобразить в немодальном окне, и это позволит продолжить работу программы, а инструкцию всегда держать на экране.

Для скрытия формы, отображенной в виде немодального окна, можно использовать метод `Hide`. Скрытие элемента управления или формы эквивалентно заданию для свойства `Visible` значения `false`. После вызова метода `Hide` свойство `Visible` возвращает значение `false` до тех пор, пока не будет вызван метод `Show`. После закрытия формы немодального окна методом `Close` свойство `DialogResult` может не вернуть значение, назначенное форме, поскольку ресурсы формы автоматически освобождаются, когда форма закрывается.

# 26. ПРИЛОЖЕНИЯ С ИНТЕРФЕЙСОМ MDI.
# 26.1. Общие сведения.
Немодальная форма — это еще не `MDI-форма`. `MDI-приложения` позволяют отображать сразу несколько документов одновременно, каждый в своем окне. Такие окна являются окнами немодальных диалогов. Это означает, что пользователь может переключаться с одного немодального окна на другое, то есть с одного документа на другой, не закрывая окна. __В `MDI`-приложении__ родительская форма должна выступать в роли контейнера для дочерних `MDI-форм`. Для этого необходимо у родительской формы установить свойство `isMdiContainer` в значение `true`. __Как правило__, у родительской формы разработчики создают меню «Окно», в который для каждого открываемого `MDI-окна` будет автоматически добавляться подпункт для перехода к этому окну, когда оно не имеет фокус. Таким образом, меню «Окно» будет содержать список всех открытых `MDI-окон`. __Новая `MDI-форма`__ добавляется в проект как обычная форма. Для `MDI-приложения`, как правило, конструируется только одна форма, на основе которой создается несколько дочерних окон — по одной для каждого документа. __Если модальное `SDI-окно`__ создается методом `ShowDialog`, то немодальное `MDI-окно` создается методом `Show`. Чаще всего элементом управления, в обработчике которого создают `MDI-окно`, является пункт меню.

----------------

__Пример__. Создание дочернего `MDI`-окна по команде меню с именем `menuNew`.

    private void menuNew_Click(object sender, EventArgs e)
    {
        Form2 wnd = new Form2();
        wnd.MdiParent = this;
        wnd.Show();
    }
Немодальное окно может отображаться либо в границах родительского окна, либо произвольным образом. Для того чтобы дочернее немодальное окно отображалось в границах родительского окна, необходимо у дочернего окна установить свойство `MdiParent`, равным ссылке на родительскую форму. Это свойство отсутствует в окне свойств `Visual Studio`. Если дочернее окно выходит за рамки родительского, то выходящая часть не отображается, а для просмотра дочернего окна в родительском создаются полосы прокрутки. По этой причине рекомендуется родительское окно распахивать на весь экран (`WindowState = Maximized`). Если родительская форма не определена как контейнер (`IsMdiContainer = false`), то свойство `MdiParent` использовать нельзя. Дочерние `MDI-окна` не закрывают родительские элементы управления, располагаясь под ними, поэтому эти элементы управления всегда доступны и поэтому в родительском окне рекомендуется использовать только меню, панель инструментов и строку состояния. Итак, основным признаком `MDI-формы` является установка её свойства `MdiParent`, равным ссылке на родительскую форму, которая определяется как `MDI-контейнер` (`IsMdiContainer = true`).
