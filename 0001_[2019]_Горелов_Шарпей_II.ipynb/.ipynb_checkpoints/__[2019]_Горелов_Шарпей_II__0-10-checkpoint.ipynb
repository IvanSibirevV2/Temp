# ПРЕДИСЛОВИЕ
Во втором томе учебника рассмотрена структурная и функциональная организация элемента управления `DataGridView`. Этот элемент отображает данные в табличном виде. Данные могут использоваться как из внешнего источника в режиме привязки, так и формироваться в программе в процессе её выполнения. Элемент управления `DataGridView` имеет богатую функциональность и состоит из множества классов. Одних только событий предоставляется более 180. Понятно, что описать все возможности этого элемента в рамках одного пособия не представляется возможным. Поэтому в данном учебнике показано, как можно использовать элемент управления `DataGridView` на примере приложения, последовательно разрабатываемого вместе с читателем. С решением очередной задачи программа становится все совершеннее и сложнее. В учебнике автор предпринял попытку решить большинство из тех задач, с которыми сталкиваются разработчики. Итогом такого подхода явилась готовая программа, которая может служить примером и основой для разработки других приложений.

# 1. СПОСОБЫ ТАБЛИЧНОГО ОТОБРАЖЕНИЯ ДАННЫХ
Очень часто информационная модель предметной области в приложении реализуется в виде коллекции объсктов. Такой коллекцией может быть список `List<T>` или `BindigList<T>`, а также таблица `DataTable` набора данных `DataSet`, загруженная из реляционной базы данных. Поскольку коллекция содержит список однотипных объектов, то отобразить такой список целесообразнее всего в табличном виде, где на каждое свойство объскта создается свой столбец. Реализовать табличное отображение можно разными способами. Выбор способа зависит от того, как пользователь будет работать со строками таблицы: только просматривать или просматривать и изменять данные. Первый способ является тривиальным и для отображения данных подходят как многострочный элемент управления TextBox, так и более сложный элемент `ListBox`. Однако такие элементы управления не создают столбцы с заголовками. Разработчик должен вручную выравнивать данные в строках, чтобы они образовывали выровненные по вертикали столбцы. Чтобы столбцы были выровнены по вертикали, необходимо выбрать моноширинный шрифт (например `Lucida Console`) и применить форматированис, указав ширину поля, например: `textBox1.Text += string.Format("{0, -15} группа {1},дисциплина {2, -18} оценка {3}\r\n ", A, B, C, D );` бразить с помощью надписей `Label`.
