# 1. СПОСОБЫ ТАБЛИЧНОГО ОТОБРАЖЕНИЯ ДАННЫХ
Очень часто информационная модель предметной области в приложении реализуется в виде коллекции объсктов. Такой коллекцией может быть список `List<T>` или `BindigList<T>`, а также таблица `DataTable` набора данных `DataSet`, загруженная из реляционной базы данных. Поскольку коллекция содержит список однотипных объектов, то отобразить такой список целесообразнее всего в табличном виде, где на каждое свойство объскта создается свой столбец. Реализовать табличное отображение можно разными способами. Выбор способа зависит от того, как пользователь будет работать со строками таблицы: только просматривать или просматривать и изменять данные. Первый способ является тривиальным и для отображения данных подходят как многострочный элемент управления TextBox, так и более сложный элемент `ListBox`. Однако такие элементы управления не создают столбцы с заголовками. Разработчик должен вручную выравнивать данные в строках, чтобы они образовывали выровненные по вертикали столбцы. Чтобы столбцы были выровнены по вертикали, необходимо выбрать моноширинный шрифт (например `Lucida Console`) и применить форматированис, указав ширину поля, например: `textBox1.Text += string.Format("{0, -15} группа {1},дисциплина {2, -18} оценка {3}\r\n ", A, B, C, D );` бразить с помощью надписей `Label`. Сушествует элемент управления ListView, который специально создан для вывода данных в виде таблицы (см. рис. 1.1). Однако этот элемени обладает сушественным недостатком: ето нельзя привязать к источнику данных, поэтому строки приходится создавать и добавлять вручную.

Рис. 1.1. Отображение данных в элементе управления `ListView`

Мы не будем изучать элемент `ListView`, так как его возможности в рассматриваемом контексте полностью перекрываются элементом управления `DataGridView`. Причем `DataGridView` не содержит недостатков элемента `ListView`. __Поскольку понятие__ «таблица» в программировании в первую очерель относится к таблицам реляционной базы данных, то `DataGridView` часто называют сеткой. Тем не менее, в данном разделе для простоты мы будем называть `DataGridView` таблицей и только там, где речь будет идти о таблицах базы данных, будем переходить к термину «сетка». __Элемент управления__ `DataGridView` является одним из самых мошных и сложных элементов управления. Он предназначен для отображения данных, хранящихся в самой таблице, или из внешнего источника данных: базы данных, XML-файла, веб-службы XML, коллекции объектов или другого источника данных. Информация отображается в традиционном формате — «строки-столбцы».

__В элемент управления__ `DataGridView` данные могут быть загружены несколькими способами:
- 1. Таблицу `DataGridView` можно привязать к источнику данных. В этом случае столбцы будут созданы автоматически и в них отобразятся привязанные данные. Такие столбцы называются привязанными.
- 2. Таблицу `DataGridView` можно заполнить данными программным способом, не используя привязку. Разработчик должен тем или иным способом создать столбцы и строки с данными. Такие столбцы называются свободными и или не привязаными.
- 3. Таблица `DataGridView` может содержать как привязанные, так и свободные столбцы. Это комбинированный режим.
- 4. С целью экономии оперативной памяти для отображения данных в таблице `DataGridView` может быть использован так называемый виртуальный режим.

В данном томе учебника упор делается на разработку программ с использованием таблиц, привязанных к источнику данных. Другие режимы так же будут рассмотрены, но не так подробно.

### 2.1. Общие сведения
Здесь и далее, до особых указаний, будем рассматривать первый способ, то есть отображение в таблице записей (объектов) из источника данных. Привязка таблицы к источнику данных значительно облегчает работу с таблицей. Для привязки используется компонент BindingSource (привязка рассмотрена в разделе 18 тома 1). __Особенность привязки состоит__ в том, что данные из внешнего источника не хранятся в ячейках таблицы. Таблица хранит лишь ссылки на объекты источника данных. Ячейки служат для вызова метода `get` свойств источника данных с целью получения и отображения данных и вызова метода `set` для записи значений в источник. __Таблица `DataGridView`__ обладает богатой базовой функциональностью, доступной по умолчанию с помощью её интерфейса. К таким функциям, в первую очередь, относятся: добавление новых строк, удаление одной или нескольких строк, изменение значений в ячейках. Кроме того, при помощи `DataGridView` можно настроить заголовки и ширину столбцов, внешний вид отдельных ячеек, зафиксировать строки и столбцы на своем месте при прокрутке таблицы, отображать элементы управления внутри ячеек, непользовать свободные столбцы наряду с привязанными, менять местами столбцы и г.д. Благодаря этому, сели не требуется выполнять специальную проверку и обработку табличных данных, использование в программе таблицы `DataGridView` потребует от разработчика минимум ручного программировання.

# ОТОБРАЖАЕМЫЕ СВОЙСТВА ТАБЛИЦЫ
В окне свойств таблицы отображается около 70 свойств. Опиать все эти свойства в одном разделе учебника не представляется возможным, поэтому рассмотрим наиболее значимые свойства.

__Anchor__ - возвращает или задает границы контейнера (формы), с которым связана таблица, и определяет способ изменения размеров таблицы при изменении размеров его родительского элемента (формы). Свойство получает комбинацию значений перечисления `AnchorStyles`. Если таблица должна быть привязана только к правой границе формы, то следует установить: `dataGridView1.Anchor = AnchorStyles. Right;`. В этом случае при перемещении правой границы формы с помощью мыши, также будет перемещаться и таблица, сохраняя расстояние до правой границы формы неизменным. Если же таблицу дополнительно привязать и к левой границе: `dataGridView1.Anchor = AnchorStyles.Left | AnchorStyles. Right;` то при изменении ширины формы расстояние как слева, так и справа от таблицы до вертикальных границ формы будет оставаться неизменным, благодаря чему таблица также будет либо растягиваться, либо сужаться. Растягивая таблицу, можно показать столбцы, оставшиеся за пределами отображаемой части. Такая установка свойства `Anchor` позволяет отказаться от горизонтальной полосы прокрутки таблицы.

__Аналогично__, только в вертикальном направлении, работают и значения `AnchorStyles.Top` и `AnchorStyles.Bottom`. Если между таблицей и границей находятся какие-либо элементы управления, то, чтобы таблица не перекрывала их при растягивании формы, рекомендустся эти влементы управления принявк, к драницам контейнера (формы) так, чтобы они перемеланись, но не распятивались. Например, для кнопки `button1`. находящейся справа ол таблицы (ем. рне. 2.4), следует установить `buttoni1.Anchor=AnchorStyles.Top | AnchorStyles.Right;`

__AutoSizeColumnsMode__ - возрашает или задает значение, указываясе, как по умолчаник, определяется ширина каждого столбца таблицы. Если ширина всех или многих столбцов таблицы должна определяться одним и тем же способом, то лучше этог способ определить в свойстве таблицы `AutoSizeColumnsMode`. Позже для конкретных столбцов можно изменить заданное по умолчанию) значение, воспользовавшись их свойством `DataGridViewColumn.AutoSizeMode`.



Свойство `AutoSizeColumnsMode` имеет значение типа `DataGridViewAutoSizeColumnsMode`, которое является перечислением и определясь следующие члены:
- __None__ - ширина столбнов актоматически не изменяется.
- __AllCell__ - ширина столбцов изменяется так, чтобы вместипь содержимос всех яческ столбцов (отображаемых и неотображаемых), включая ячейки заголовков.
- __AllCellsExceptHeader__ -  ширина столбцов изменяется так, что бы вместиль содержимое всех яческ столбцов, за исклюочением яческ заголовков.
- __DisplayedCellis__ -  ширина столбцов изменяется так, чтобы вместить, содержимое всех ячеек столбцов, которые находятся в строках, отображаюшихся на экране в настоящий момент, включая ячейки заголовков. Недостаком данного способа является то, что ширина каждого столбца вычисляется только на основе видимой части таблицы в момент ее первого отображения. Это означает, что, ссли в результате вертикальной прокрутки таблицы встретится ячейка с более длинным содержимым, то оно будет усечено, а к видимой части содержимого будут добавлены три точки. Однако как только пользователь перейдет в режим заполнения этой ячейки, ширина столбца изменится в соответствии с введенным значением.
- __DisplayedCellsExceptHeader__ — ширина столбцов изменяется так, чтобы вместить содержимое всех ячеек столбцов, которые находятся в строках, отображающихся на экране в настоящий момент, исключая ячейки заголовков. Данный режим имеет те же недостатки, что и `DisplayedCells`.
- __ColumnHeader__ - ширина столбцов изменяется так, чтобы вместить содержимое ячеек заголовков столбцов.
- __Fill__ — ширина столбцов подбирается таким образом, чтобы суммарная ширина всех столбцов в точности заполняла отображаемую область элемента управления, а прокрутка по горизонтали требовалась только для того, чтобы не допускать уменьшения ширины столбцов ниже значений свойства `DataGridViewColumn.MinimumWidth`, которое по умолчанию равно 5. Относительная ширина столбцов определяется значениями свойства `DataGridViewColumn.FillWeight`.

# 2.4. Архитектура DataGridView
### Особенности привязки DataGridView к данным
Привязка таблицы `DataGridView` к источнику данных имеет ряд особенностей. Так, `DataGridView`, в отличие от элементов управления `ListBox` и `ComboBox`, позволяет, используя интерфейс таблицы, вносить изменения в таблицу: удалять строки, добавлять новые записи и изменять содержимое яческ. Такая особенность обусловлена тем, что в режиме привязки данных все значения ячеек для столбцов, связанных с данными, хранятся во внешнем источнике данных. Это означает, что считывание значе ния из ячейки таблицы осуществляется путем выполнения метода `get` свойства, соответствующего ячсикс, а изменсние содержимого ячейки (любым способом) приводит к вызову метода `set` свойства. Программным способом добавлять новые записи в привязанную к источнику данных таблицу нельзя. Поскольку компонент `BindingSource` обеспечивает двустороннюю привязку, то такие операции, как добавление, удаление или изменение записен, могут выполняться или в источнике данных, или в таблице (программным способом или через интерфейс таблицы).

# Реальное и отображаемое значения ячейки.
В режиме привязки каждая ячейка связана с соответствующим ей свойством источника данных. Для доступа к значению ячейки используются два свойства: `Value` и `FormattedValue`. Свойство `Value` ячейки получаст или возвращает значение соответствущего свойства источника данных. Такое значение может иметь тип `int`, `float`, `DateTime`, `string` и т.д. Свойство `FormattedValue` возвращает отформатированное значение, предоставленное свойством `Value`, для его отображения. Ячейка имеет тип, получаемый от столбца, или назначенный разработчиком. По аналогии с типами столбцов, используются следующие типы ячеек: __Во-первых__, `DataGridViewButtonCell`; __Во-вторых__, `DataGridViewCheckBoxCell`; __В-третьих__, `DataGridViewComboBoxCell`; __в четвёртых__, `DataGridViewHeaderCell`; __в пятых__, `DataGridViewImageCell`; __в шестых__, `DataGridViewLinkCell`; __В седьмых__, `DataGridViewTextBoxCell`. Все перечисленные типы являются наследниками типа `DataGridViewCell`.

Наиболее распространенным является тип `DataGridViewTextBoxCell`, используемый для отображения текста. То есть каждое значение свойств источника данных для отображения в таблице преобразуется в строку типа `String`. Таким образом, в режиме привязки ячейки таблицы не хранят никаких значений. Единственно, что хранит таблица, это ссылки на объекты источника данных. __Для того чтобы понять__, о каком значении идет речь, будем использовать следующие термины: __«реальное значение ячейки»__ (или просто «значение ячейки») — это значение из источника данных, которое получает или возвращает свойство ячейки `Value`; __«отформатированное значение ячейки»__ (или «отображаемое значение ячейки») — это значение, которое возвращает свойство ячейки `FormattedValue`. __Изменение данных в таблице__. Когда пользователь изменяет значение в ячейке, это значение изменяется непосредственно в источнике данных. Если пользователь создает новый объект с помощью строки для добавления новой записи, новый объект создается конструктором по умолчанию в источнике данных до того, как эта запись будет заполнена. Удаление записи из таблицы, естественно, приводит к удалению соответствующего объекта из источника данных. __Для изменения__ содержимого ячеек используются соответствующие методы `set` свойств объектов источника. Никаких обновлений привязки в перечисленных изменения не требуется и не выполняется.

__Изменение источника данных путем добавления или удалення записей__. Изменение источника данных не всегда приводи к автоматическому отображению измененных данных в таблице. Коллекция `List<T>` не уведомляет компонент `BindingSource` о выполненных в списке изменениях. Поэтому, если не обновить приказку, будет наблюдаться несоответствие отображаемого значения реальному. Например, вставим на место второй записи в список `List<T>` новый объект. Таблица будет продолжать отображать старое состояние. Однако, если программным способом прочитать какую-либо ячейку второй записи таблицы, то это будет ячейка другой записи — новой. С удалением записи дело обстоит еще хуже: после удаления записи из списка, количество строк в таблице останется прежним, а в списке все записи, располагающиеся после удаленной, сдвинутея на одну запись к началу списка, то есть записей станет на одну меньше. Таблица попытается прочитать последнюю запись в списке, которой нет, и выдаст серию сообщений об отсутствии последней записи таблицы. Поэтому после добавления записи в список или удаления записи из списка следует тут же обновить таблицу, выполнив оператор `bindingSource.ResetBindings(false);`. Коллекция `BindingList<T>` уведомляет о добавлении новых записей (`Add`, `AddRange`, `Insert`) и об удалении записей (`Remove`, `RemoveAt`), что приводит к перерисовке таблицы.

__Изменение значения отдельного свойства источника данных__. В пользовательском классе разработчики, как правило, не реализуют уведомление связывателя об изменении значений свойств. После изменения какого-либо значения свойства в источнике в таблице по-прежнему будет отображаться старое значение. Однако попытка прочитать ячейку, соответствующую измененному свойству, приведет к чтению нового значения. Связано это с тем, что каждое значение ячейки читается из источника данных. Чтобы таблица вместо старого значения ячейки отобразило новое, разработчик должен перерисовать строку или таблицу. Для этого может быть использован один из следующих методов:
- __ResetItem__ - вызывает в таблице, привязанной к компонентy `BindingSource`, повторное считывание элемента источника данных по указанному индексу и обновление его отображаемого значения.
- __ResetBindings__ — вызывает в таблице, привязанной к компоненту `BindingSource`, повторное считывание всех элементов источника данных и обновление их отображаемых значений.

Как Вы видите, можно перерисовать одну строку, а не всю таблицу, что предпочтительней с точки зрения производительности. Таким образом, если Вы реализуете алгоритмы добавления или удаления объектов в источнике данных, то используйте сразу же после добавления или удаления метод `ResetBindings`. Если изменяете отдельное значение - `ResetItem`. Если Вы выполняете изменения данных в таблице, дополнительно обновлять отображение ячеек и строк не надо — таблица это сделает за Вас. Какие открытые свойства не отображаются в таблице. Как было отмечено выше, наиболее распространенным типом является `DataGridViewTextBoxColumn`. Для преобразования в текст числовых данных, данных типа `DateTime`, символов и т.д. применяется метод `ToString`. Если класс содержит открытое свойство пользовательского типа, например `public UserClass User { get; set; }`, то в результате привязки столбец для этого свойства создастся, а для отображения значений в столбце будет вызываться метод `ToString` этого типа (то есть само свойство в строку не преобразуется, что естественно). Благодаря тому, что столбец для пользовательского типа создается, можно преобразовать его в тип `DataGridViewComboBoxColumn`, о чем будет рассказано позже. В таблице, привязанной к источнику данных, не для всех открытых свойств создаются столбцы, Например, для следующего свойства столбец создан не будет: `public List< UserClass > User { get; set; }`

# Коллекции таблицы DataGridView

# ОСОБЕННОСТИ СОЗДАНИЯ НОВОЙ ЗАПИСИ
__После того как пользователь__ переместит фокус ввода на строку для добавления новой записи, возможны следующие варианты завершения работы с этой строкой: __Во-первых__, Если пользователь ввел с клавиатуры первый символ в какую-либо ячейку новой строки, фиксируется режим редактирования ячейки. В этом случае создается новая запись в источнике данных; ниже вводимой записи появляется очередная строка для добавления новой записи. Если после заполнения ячеек данными пользователь переместит фокус на любую предшествующую или очередную новую строку, новая запись будет зафиксирована как созданная. Новую запись будем называть зафиксированной, если её добавление нельзя будет отменить клавишей `Escape`. __Во-вторых__, Пользователь не стал редактировать ни одной ячейки (не вошел в режим редактирования) и возвратился на любую предшествующую строку, переместив курсор или нажав `Escape`. В этом случае новая запись создана не будет. __в-третьих__, Как только пользователь впервые входит в режим редактирования какой-либо ячейки, таблица создает новую запись в источнике данных; ниже появляется очередная строка для добавления новой записи. Пользователь заполняет какие-либо ячейки данными, но потом отказывается от создания записи, нажимая клавишу `Escape`. Первое нажатие отменяет редактирование текущей ячейки, а второе приводит к удалению записи из источника данных и перемещению фокуса ввода на предыдущую строку.

__Важное замечание__: при отказе от добавления новой записи с помощью клавиши `Esc` (вариант 2 или 3), осуществляется повторное чтение последней записи коллекции и её отображение в таблице. Если разработчик для отображения записей, соответствующих некоторому критерию, использует свойство `Row.Visible` (см. раздел 2. 12), то этой последней записью может оказаться запись с другим критерием. Возникает некорректная ситуация. Исправить эту ситуацию можно только нетривиальным способом, поэтому лучше отказаться от добавления новой записи с помощью интерфейса таблицы и использовать для добавления отдельную форму (см. раздел 2.12). Тем не менее, в комплексном примере, приведенном в разделе 2.17, показан способ решения этой проблемы (см. событие `RowPostPaint`). ледует заметить также, что независимо от того, создана новая запись или нет, любое перемещение со строки для новой записи на другую строку приводит к появлению события `RowValidated`.

# КОЛЛЕКЦИЯ ЯЧЕЕК
Ячейки образуют коллекцию ячеек строки. Ячейка является объектом типа `DataGridViewCell`, имеющим множество свойств. Наиболее значимыми свойствами являются следующие:
- __Value__ — получает или задает значение, связанное с этой ячейкой. Это реальное значение ячейки (свойства источника данных), а не отображаемое в ячейке. Например, реальным значением может быть номер вуза, а отображаемым — его название. К сожалению, это значение имеет тип `object` и его необходимо приводить к типу соответствующего свойства источника данных.
- __ValueType__ - получает или задает тип данных для значений в ячейке. Это свойство, как правило, возвращает тип данных источника привязки.
- __FormattedValue__ - получает значение ячейки, отформа тированное для отображения. Это значение, как правило, можно привести к типу `string`.
- __ColumnIndex__ — получает индекс столбца для данной ячейки.
- __RowIndex__ — получает индекс родительской строки ячейки.
- __Style__ - получает или задает стиль ячейки.
- __InheritedStyle__ — получает стиль, применяемый в текущий момент к ячейке. Элемент управления `DataGridView` отображает ячейки, используя стили, указываемые свойством `InheritedStyle` ячейки, которое наследует стили из других свойств типа `DataGridViewCellStyle`.

Стили, задаваемые свойством `Style`, переопределяют стили, задаваемые всеми другими свойствами стиля ячейки, но не обязательно показывают все стили, которыми определяется внешний вид ячейки.
- __ReadOnly__ - получает или задает значение, показывающее, можно ли изменять данные в ячейке.

# ВЫБРАННЫЕ СТРОКИ, СТОЛБЦЫ И ЯЧЕЙКИ
Таблица `DataGridView` поддерживает понятие «выбранная коллекция объектов» и «текущий объект». Пользователь может выбрать несколько строк, столбцов или ячеек, если свойство `DataGridView.MultiSelect` равно `true`. Это стандартное значение. На практике выборка коллекции строк чаще всего используется для их одновременного удаления. В остальных случаях, как правило, используется только первая строка или столбец коллекции. Свойство `DataGridView.SelectionMode` возвращает или задает значение, указывающее, каким образом могут быть выбраны ячейки объекта `DataGridView`. Возможные значения задает перечисление `DataGridViewSelectionMode`:
- __CellSelect__ - можно выбрать одну или несколько отдельных ячеек. Строку или столбец выбирать нельзя.
- __FullRowSelect__ - вся строка выбирается путем щелка заголовка строки или ячейки, содержащейся в этой строке. Выбрать отдельную ячейку нельзя.
- __RowHeaderSelect__ - строка может быть выбрана путем щелчка заголовка строки. Отдельную ячейку можно выбрать, щелкнув ее.

# 2.7. Проблемы редактирования ячеек
При редактировании ячеек существующей или добавляемой записи возникает ряд проблем, связанных с вводом данных по определенному формату и проверкой вводимых значений.
- __Во-первых__, вводимое значение должно соответствовать типу свойства источника данных, которое должно отображаться в ячейке. Например, нельзя в ячейку вводить текст, не являющийся числом, если свойство, соответствующее ячейке, имеет числовой тип. В ячейку для свойства `DateTime` можно вводить только дату и/или время.
- __Во-вторых__, разные ячейки одной записи могут иметь зависимые значения. Например, дата приезда не может больше даты отьезда. Если тип вводимого значения можно проверить в момент его ввода, то значения зависимых ячеек можно проверить только после завершения редактирования строки. Связано это с тем, что разработчик во время редактирования отдельной ячейки не может вынести суждение о том, что редактирование зависимых ячеек было завершено.
- __В-третьих__, иногда необходима подмена реального значения другим, отображаемым. Например, если число 0, свидетельствует об отсутствии значения, то лучше отобразить в ячейке пустое значение.
- __В-четвертых__, для ввода сложных значений, например даты, целесообразно использовать специальный элемент управления, например `MonthCalendar`, который желательно отображать рядом с редактируемой ячейкой. Использование элементов управления для ввода значений позволяет получить одинаковое представление вводимых данных.
- __В-пятых__, удаление выделенных строк пользователем осуществляется без дополнительного разрешения и безвозвратно. Здесь перечислена только часть проблем. Ниже будет показано, как с использованием обработчиков различных событий таблицы решаются перечисленные выше проблемы.

# 2.8. Проверка значений, введенных в ячейки
### ОБЩИЕ СВЕДЕНИЯ
Проверка значений ячеек таблицы, является одной из основных задач, решаемых разработчиком. Эта задача разбивается на следующие подзадачи: __во-первых__. Проверка корректности значений, вводимых в ячейки. __во-вторых__. Проверка ячеек на отсутствие значений. __в-третьих__. Проверка согласованности зависимых ячеек одной строки. __Большинство задач__, связанных с таблицей, решается с помощью обработки событий таблицы. Число этих событий, как уже было отмечено, превышает 180. Ниже будут рассмотрены только события, используемые нами для решения наиболее распространенных задач. __Ввеленное значение может__ быть некорректным по причине его несоответствия типу исходного свойства. Например, если тип свойства `uint`, то вводить в ячейку можно только положительные целые числа. Перед записью значения в источник данных специальный объект, связанный с компонентом `BindingSource`, автоматически проверяет корректность введенного значения (значение преобразуется с помощью метода `Parse`) и в случае обнаружения ошибки генерирует событие `DataError`. Если программа не перехватывает это событие, то система выдает окно с информацией об ошибке. Такое окно слишком перегружено малопонятной пользователю информацией, поэтому разработчики иногда сами обрабатывают событие `DataError`, а чаще всего, делают проверку ячейки до её анализа в `BindingSource` и устраняют все ошибки.

Для проверки введенного в ячейку значения лучше всего подходит событие `CellValidating`. Это событие происходит, когда фокус ввода перемещается из текущей ячейки в любую другую. На уровне выполнения кода такое поведение реализуется путем отмены события `CellValidating`. Когда это событие отменяется в режиме привязки данных, новое значение не отправляется принудительно в базовый источник данных. Каждая строка и ячейка таблицы имеют свойство `ErrorText`. Если этому свойству присвоить какое-либо значение, например

то в заголовке строки появится значок ошибки (см. рис. 2.14). Сообщение появляется тогда, когда пользователь наводит курсор на значок ошибки. Чтобы убрать из заголовка строки этот значок, необходимо очистить свойство `ErrorText`:

# ПРОВЕРКА ЗНАЧЕНИЙ НА УРОВНЕ СТРОКИ
__Событие `RowValidating`__. Как было отмечено выше, событие `CellValidating` не годится для проверки зависимых ячеек. Необходимо событие, предоставляющее в распоряжение разработчика строку таблицы целиком после завершения всех изменений. Таким событием является `RowValidating`. По своей функциональности это событие похоже на событие `CellValidating`, рассмотренное выше. Отличие состоит в том, что, если событие `CellValidating` происходит для ячейки, которая теряет фокус ввода, то событие `RowValidating` происходит для строки, которая теряет фокус ввода. Благодаря этому, данное событие можно использовать для выполнения проверки всех значений в строке.

__Обработчик события `RowValidating`__. Рассмотрим параметры обработчика события `RowValidating` на примере таблицы `dataGridView1`. Формат заголовка:

    dataGridView1_RowValidating (object sender, DataGridViewCellCancelEventArgs e)
Параметр `е` имеет следующие свойства:
- __ColumnIndex__ — индекс столбца ячейки в текущей строке, с которой осушествляется переход.
- __Rowindex__ — индекс текущей строки.
- __Cancel__ — значение `true`, если событие следует отменить, в противном случае - значение `false`. Значение `true` не позволит пользователю оставить текущую строку неотредактированной и перейти на другую строку. При входе в обработчик свойство `Cancel` по умолчанию имеет значение `false`.

Поскольку в распоряжении разработчика находятся все ячейки, он может проверить каждую по отдельности и в сочетании друг с другом. Казалось бы, имея событие `RowValidating`, можноcотказаться от проверки ячеек в обработчике события `CelIValidating`. Однако это не так. Если разработчик не отреагирует на ошибочные значения в обработчике `CellValidating`, то на них отреагирует система возбуждением события `DataError` и выводом окна с описание ошибки. Событие `DataError` происходит после `CellValidating`, но до `RowValidating`. Следуст заметить: при переходе фокуса на другую строку управление также получает и обработчик события `CellValidating`.

__Особенности работы с новой строкой__. Событие `RowValidating` происходит так же и при добавлении новой записи, когда пользователь перемещает фокус из новой строки в любую другую. Если строка не прошла проверку, задайте `Cancel-true`. Отмена этого события запрещает пользователю выход из недопустимой строки и предотвращает ее сохранение во внешнем источнике данных в режиме привязки данных. Отказаться от добавления новой записи после перехода на новую строку, но до вхождения в режим редактирования ячейки можно просто возвратившись с новой строки на любую старую строку или нажав клавишу `Escape`. В этом случае новая запись не добавляется в источник данных, но, тем не менее, возникаст событие `RowValidating`, обработка которого для этой ситуации часто приводит к ошибкам, так как ячейки строки для новой записи имеют значение `null`. __Если пользователь начал__ редактировать какую-либо ячейку новой записи, запись сразу же добавляется в источник данных, но пользователь (пока запись не потеряла фокус) может отказаться от этой записи. Для этого необходимо нажать два раза клавишу `Escape`. Первый раз - для восстановления исходного значения текущей ячейки (и выходу из режима редактирования) и второй раз - для отказа от новой записи. В этом случае система вернет строку для новой записи в исходное состояние (все значения равны `null`), а новая запись будет автоматически удалена системой из источника с помощью метода `Remove`. К сожалению, этот метод выполняется после возникновения события `RowValidating`, обра ботка которого для этой ситуации, часто приводит к ошибкам, так как запись еще не удалена, а ячейки новой строки, соответствующей записи, имеют значение `null`.
